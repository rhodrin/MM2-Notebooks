# Complex numbers

Complex numbers are represented in the form

\begin{equation*}
 z=x+iy,
\end{equation*}

where $i=\sqrt{-1}$ (hence $i^2=-1$) and $x=\mathrm{Re}(z)$ is the *real part* of the complex number and $y=\mathrm{Im}(z)$ is its imaginary part. The *complex conjugate* of any complex number is defined by

\begin{equation*}
 \bar{z}=x-iy,
\end{equation*}

and hence

\begin{align*}
 x&=\frac{z+\bar{z}}{2}, \\
 y&=\frac{z-\bar{z}}{2i}.
\end{align*}

Complex numbers can also be represented in polar form

\begin{equation*}
 z(=x+iy)=re^{i\theta}=r(\cos(\theta)+i\sin(\theta)),
\end{equation*}

where

\begin{align*}
 r&=|z|=(z\bar{z})^{1/2}, \\
 \tan(\theta)&=\frac{\mathrm{Im}(z)}{\mathrm{Re}(z)}.
\end{align*}

We can plot complex numbers in the complex plane in a similar manner to how we plot vectors in $\mathbb{R}^2$. Below we plot $z=-2+5i$:

```python
# Define a plotter function:

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def argand(z, xl=-1.2, xr=1.2, yl=-1.2, yr=1.2):

    rc = {"xtick.direction" : "inout", "ytick.direction" : "inout",
          "xtick.major.size" : 5, "ytick.major.size" : 5,}
    with plt.rc_context(rc):
        fig, ax = plt.subplots(figsize=(6,4), dpi= 120, facecolor='w', edgecolor='k')
        ax.scatter(z.real,z.imag)

        ax.spines['left'].set_position('zero')
        ax.spines['right'].set_visible(False)
        ax.spines['bottom'].set_position('zero')
        ax.spines['top'].set_visible(False)
        ax.xaxis.set_ticks_position('bottom')
        ax.yaxis.set_ticks_position('left')
        ax.set_xlim([xl, xr])
        ax.set_ylim([yl, yr])
        ax.set_aspect('equal')
        ax.set_xlabel('Re(z)', loc='right')
        ax.set_ylabel('Im(z)', loc='top')

        # make arrows
        ax.plot((1), (0), ls="", marker=">", ms=10, color="k",
               transform=ax.get_yaxis_transform(), clip_on=False)
        ax.plot((0), (1), ls="", marker="^", ms=10, color="k",
                transform=ax.get_xaxis_transform(), clip_on=False)

        plt.show()
```

```python
z=-2+5j
argand(z, xl=-5.5, xr=5.5, yl=-5.5, yr=5.5)
```

Using complex numbers we can form *complex equations* which essentially represent two simultaneous in $x$ and $y$ compactly. For example,

\begin{equation*}
 2z+\frac{\bar{z}}{1-i}=-4+2i.
\end{equation*}

We can re-write this as

\begin{align*}
 2z+\frac{\bar{z}(i+1)}{2}&=-4+2i, \\
 4z+\bar{z}+i\bar{z}&=-8+4i, \\
 (5x+y)+i(x+3y)&=-8+4i.
\end{align*}

Taking the real and imaginary parts we have two simultaneous in $x$ and $y$:

\begin{align*}
 5x+y&=-8, \\
 x+3y&=4,
\end{align*}

which can be solved as normal to give $x=-2$ and $y=2$ and hence

\begin{equation*}
 z=-2+2i,
\end{equation*}

Solves our original equation.

From the definitions above we can derive some important **identities:**

**Euler's identity**

\begin{equation*}
 e^{i\pi}+1=0,
\end{equation*}

**de Moivre's identity**

\begin{equation*}
 (\cos\theta + i\sin\theta)^n=\cos n\theta + i\sin n\theta,
\end{equation*}

and under the restriction that $|z|=1$ the following definitions for trigonometric functions can be derived:

\begin{align*}
 \cos(n\theta)=\frac{1}{2}\left(z^n+z^{-n}\right), \\
 \sin(n\theta)=\frac{1}{2i}\left(z^n-z^{-n}\right).
\end{align*}

### Example, coursework 1 problem 5:

**Use de Moivre's theorem to find an expression for $\sin^4\theta$ in terms of $\cos m\theta$.**

From de Moivre's theorem we can show that

\begin{equation*}
 \sin(n\theta)=\frac{1}{2i}\left(z^n-z^{-n}\right),
\end{equation*}

when $|z|=1$. Hence

\begin{equation*}
 \sin^4(\theta)=\left(\frac{1}{2i}\right)^4\left(z-\frac{1}{z}\right)^4.
\end{equation*}

Expanding the right hand side, collecting terms and utilizing the above identity we arrive at

\begin{equation*}
 \sin^4\theta=\frac{1}{8}\cos(4\theta)-\frac{1}{2}\cos(2\theta)+\frac{3}{8}.
\end{equation*}

Lets quickly test our solution:

```python
import numpy as np

# Some trial theta:
t1 = 0.235*np.pi
t2 = 5.678*np.pi

def s4t(theta):
    return 1/8*np.cos(4*theta)-0.5*np.cos(2*theta)+3/8

print(np.sin(0)**4, s4t(0))
print(np.sin(t1)**4, s4t(t1))
print(np.sin(t2)**4, s4t(t2))
```

## Roots of complex numbers

Another important topic we covered is taking $n$-th roots of complex numbers.

Above we noted that we can write $z=re^{i\theta}$ (where $r$ and $\theta$ are discussed above). Due to the periodicity of $2\pi-\theta$ this can also be written as

\begin{equation*}
 z=re^{i\theta+2m\pi},
\end{equation*}

where $m$ is any integer. The $n$-th roots of a complex number $z$ can therefore be written as:

\begin{equation*}
 z^{\frac{1}{n}}=r^{\frac{1}{n}}e^{i\left(\frac{\theta}{n}+\frac{2m\pi}{n}\right)}.
\end{equation*}

Each $2\pi$ window will then contain $n$ roots which are uniquely located in the complex plane (i.e. any other roots will differ by an argument of $2l\pi$ for some integer $l$).

### Example

**Compute the cubic roots of $z=8e^{i\pi/2}$ and state solutions for the branch $-\pi\leq\theta<\pi$.**

Since

\begin{equation*}
 z^{\frac{1}{n}}=r^{\frac{1}{n}}e^{i\left(\frac{\theta}{n}+\frac{2m\pi}{n}\right)}.
\end{equation*}

we have here that

\begin{equation*}
 z^{\frac{1}{3}}=r^{\frac{1}{3}}e^{i\left(\frac{\pi}{6}+\frac{2m\pi}{3}\right)}.
\end{equation*}

The unique roots can then be found by choosing **any three neighboring** values of $m$, e.g. (1, 2, 3), (0, 1, 2) etc. will all do. Here we want the roots satisfying $-\pi\leq\theta<\pi$. By inspection, it's simple to see that this will equate to choosing $m=-1, 0, 1$ which yield the roots:

\begin{align*}
 z_{r1}&=2e^{-i\pi/2}, \\
 z_{r2}&=2e^{i\pi/6}, \\
 z_{r3}&=2e^{i5\pi/6}.
\end{align*}

Note that since these are cubic roots they are separated by an argument of $2\pi/3$.

## First order non-homogeneous ODEs

We begin this lecture by reviewing the solution procedure for ODEs of the form

\begin{equation*}
 \frac{\mathrm{d}y}{\mathrm{d}x}+f(x)y=r(x).
\end{equation*}

We showed that solutions can be written in the form

\begin{equation*}
 y=y_p+y_h
\end{equation*}

where $y_h$ is the solution to the corresponding homogeneous problem

\begin{equation*}
  \frac{\mathrm{d}y}{\mathrm{d}x}+f(x)y=0,
\end{equation*}

and can hence be written as

\begin{equation*}
  y_h=ce^{-\int f(x)\mathrm{d}x}=cv(x)
\end{equation*}

and $y_p$ is the *particular integral* given by

\begin{equation*}
  y_p=u(x)v(x)
\end{equation*}

with

\begin{equation*}
  u=\int \frac{r(x)}{v}\mathrm{d}x.
\end{equation*}

This solution procedure is often termed *the method of variation of parameters*.

### Example 1:

Solve

\begin{equation*}
 {x^2}y' + xy + 2=0.
\end{equation*}

First lets re-write this in the form

\begin{equation*}
 y' + \frac{y}{x}=-\frac{2}{x^2},
\end{equation*}

and identify

\begin{equation*}
 f(x)=\frac{1}{x}, \; \; \; \; r(x)=-\frac{2}{x^2}.
\end{equation*}

Hence

\begin{equation*}
 v(x)=e^{-\int f(x)\mathrm{d}x}=e^{-\int x^{-1}\mathrm{d}x}=e^{-\ln(x)}=\frac{1}{x},
\end{equation*}

and therefore

\begin{equation*}
 y_h=\frac{c}{x}.
\end{equation*}

Now that $v(x)$ is known we can solve for $u(x)$:

\begin{equation*}
 u=\int \frac{r(x)}{v}\mathrm{d}x=\int \frac{-2x^{-2}}{x^{-1}}\mathrm{d}x=-2\ln(x),
\end{equation*}

and hence

\begin{equation*}
 y_p=uv=-\frac{2\ln(x)}{x}.
\end{equation*}

Hence, bringing these together we have the general solution

\begin{equation*}
 y=-\frac{2\ln(x)}{x}+\frac{c}{x}
\end{equation*}

### Example 2 (Coursework 2, problem 1):

Consider the first order non-homogeneous ordinary differential equation

\begin{equation*}
 \frac{\mathrm{d}y}{\mathrm{d}x}+3y=x^2, \;\;\; y(0)=0.
\end{equation*}

As we saw in lectures, the general solution to problems of this form can be written as $y=y_p+y_h$ where
$y_h$ is the solution to the corresponding homogeneous problem and $y_p$ is the particular solution to the
non-homogeneous problem.

a) Find $y_h$.

b) Find $y_p$. Note that you will need to make use of integration by parts. Hence write down the general solution $y=y_{p}+y_{h}$.

c) Let us now instead assume $y_p$ to be a general second order polynomial, that is

\begin{equation*}
 y_p=ax^2+bx+c.
\end{equation*}

Show that combining this particular solution with our homogeneous solution from part a) we arrive at the same
general solution that was found in part b). Note: to do this we will first need substitute this form of $y_p$
into our differential equation and find $a$, $b$ and $c$ such that the equation is satisfied.

d) Using the provided initial/boundary condition $y(0)=0$, write down the specific solution to this problem.

#### Solution:

a) The equivalent homogeneous equation is

\begin{equation*}
 \frac{\mathrm{d}y_h}{\mathrm{d}x}+3y_h=0.
\end{equation*}

We can solve this via separation of variables. Utilizing this method we arrive at

\begin{align*}
 \int\frac{\mathrm{d}y_h}{y_h}&=-\int 3\mathrm{d}x, \\
 \ln y_h&=-3x+c, \\
 y_h&=Ae^{-3x},
\end{align*}

where $A$ is an unknown constant.

b) Owing to the linearity of the equation (and that it's clear we can add $0$ to each side of the equation!) we can write the general solution in the form $y=y_h+y_p$ where $y_p$ is a particular solution that satisfies the non-homogeneous part of the equation.

Using the method variation of parameters, the particular solution can be written in the form

\begin{equation*}
 y_p=v\int\frac{r(x)}{v}\mathrm{d}x,
\end{equation*}

where $v=e^{-3x}$, $r(x)=x^2$ (see above for details). 

Hence

\begin{equation*}
 y_p=e^{-3x}\int x^2 e^{-3x}\mathrm{d}x.
\end{equation*}

Evaluating this integral we arrive at

\begin{equation*}
 y_p = \frac{x^2}{3}-\frac{2x}{9}+\frac{2}{27}.
\end{equation*}

The general solution is therefore

\begin{equation*}
 y = ce^{-3x}+\frac{x^2}{3}-\frac{2x}{9}+\frac{2}{27}.
\end{equation*}

c) Instead of using the variation of parameters method, we're now using a trial solution (which is just what we did when solving some second order non-homogeneous ODEs!). Our trial solution is

\begin{equation*}
 y_p=Ax^2+Bx+C,
\end{equation*}

and therefore

\begin{equation*}
 y_p'=2Ax+B.
\end{equation*}

Upon substitution into our original equation we arrive at

\begin{equation*}
 2Ax+B+3(Ax^2+Bx+C)=x^2.
\end{equation*}

Comparing coefficients of powers of $x$ between the $LHS$ and $RHS$ we have

\begin{align*}
 x^2: &\qquad 3A=1 \rightarrow A=\frac{1}{3}, \\
 x^1: &\qquad 2A+3B=0 \rightarrow B=-\frac{2}{9}, \\
 x^0: &\qquad B+3C=0 \rightarrow C=\frac{2}{27},
\end{align*}

and hence

\begin{equation*}
 y_p = \frac{x^2}{3}-\frac{2x}{9}+\frac{2}{27},
\end{equation*}

and

\begin{equation*}
 y = ce^{-3x}+\frac{x^2}{3}-\frac{2x}{9}+\frac{2}{27}.
\end{equation*}

which is the same as we found in part b).

d) We now find the specific solution when $y(0)=0$, that is, this condition will allow us to compute $c$ in our general solution. Applying this condition we have

\begin{equation*}
 c+\frac{2}{27}=0 \rightarrow c=-\frac{2}{27},
\end{equation*}

and hence

\begin{equation*}
 y = -\frac{2}{27}e^{-3x}+\frac{x^2}{3}-\frac{2x}{9}+\frac{2}{27}.
\end{equation*}

## Linear second order ODEs

The most general form of a linear second order equation takes the form

\begin{equation*}
 \frac{\mathrm{d}^2 y}{\mathrm{d}x^2}+f(x)\frac{\mathrm{d}y}{\mathrm{d}x}+g(x)y=r(x).
\end{equation*}

First, lets review what we've learned about homogeneous equations with constant coefficients, that is, equations of the form

\begin{equation*}
 \frac{\mathrm{d}^2 y}{\mathrm{d}x^2}+b\frac{\mathrm{d}y}{\mathrm{d}x}+cy=0,
\end{equation*}

where $b$ and $c$ are real numbers.

In previous lectures we showed that this equation can be factorized as

\begin{equation*}
 \left(\frac{\mathrm{d}}{\mathrm{d}x}-\lambda_1\right)\left(\frac{\mathrm{d}}{\mathrm{d}x}-\lambda_2\right)y=0.
\end{equation*}

where

\begin{equation*}
 \lambda_{1,2}=-\frac{b}{2}\pm\frac{\sqrt{b^2-4c}}{2}.
\end{equation*}

We are then left with two first order ODEs of the form

\begin{equation*}
 \frac{\mathrm{d}y}{\mathrm{d}x}-\lambda y = 0,
\end{equation*}

to solve, which yield solutions of the form

\begin{equation*}
 y\propto e^{\lambda x}.
\end{equation*}

Hence, owing to the linearity of the underlying equation we expect general solutions of the form

\begin{equation*}
 y=Ae^{\lambda_1 x} + Be^{\lambda_2 x}.
\end{equation*}

Note that the values of $\lambda_1$ and $\lambda_2$ can be found by substituting the expected form of our solution ($e^{\lambda x}$) back into the differential equation, which gives

\begin{equation*}
 \lambda^2 e^{\lambda x} + b \lambda e^{\lambda x} +c e^{\lambda x} = 0.
\end{equation*}

Dividing through by $e^{\lambda x}$ we are left with a quadratic equation for $\lambda$:

\begin{equation*}
 \lambda^2+b\lambda+c=0,
\end{equation*}

which has solutions

\begin{equation*}
 \lambda_{1,2}=-\frac{b}{2}\pm\frac{\sqrt{b^2-4c}}{2}.
\end{equation*}

just as we showed above!

### Regimes and examples

We'll now look at what solutions PDEs of this type can yield and relate each to a *mechanical system*.

### Two distinct real roots ($b^2>4c$)

That is, the solution takes the form

\begin{equation*}
 y=Ae^{\lambda_1 x}+Be^{\lambda_2 x},
\end{equation*}

with $\lambda_1$, $\lambda_2$ both real.

**Example:**

\begin{equation*}
 \frac{\mathrm{d}^2 y}{\mathrm{d}x^2}+\frac{\mathrm{d} y}{\mathrm{d}x}-6y=0.
\end{equation*}

Using one of the methods above this yield $\lambda_1=3$, $\lambda_2=-2$ (or $\lambda_1=-2$, $\lambda_2=3$ - it doesn't matter how we label these, we'll get equivalent results). Hence

\begin{equation*}
 y=Ae^{3x}+Be^{-2x}
\end{equation*}

How do functions of this kind look? Lets pick some initial conditions (so that we can evaluate $A$ and $B$) and see. Lets take:

\begin{equation*}
 y(0)=0, \; \; \; \; \; \; \frac{\mathrm{d}y}{\mathrm{d}x}(0)=4.
\end{equation*}

Hence,

\begin{equation*}
 A+B=0, \; \; \; \; \; \; 3A-2B=4,
\end{equation*}

which gives

\begin{equation*}
 A=\frac{4}{5}, \; \; \; \; \; \; B=-\frac{4}{5}.
\end{equation*}

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

N = 1000
x = np.linspace(-10,10,N)

def y_2real(x, A, B):
    return A*np.exp(A*x)+A*np.exp(B*x)

y = y_2real(x, 0.8, -0.8)

plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.ylim((0, 10))
plt.legend()
plt.show()
```

As we saw previously, 2nd Order ODEs can also be used to model mechanical systems, one such system being an oscillating mass on a spring:

In this case the differential equation can be written as

\begin{equation*}
 m\frac{\mathrm{d}^2 y}{\mathrm{d}t^2}+c\frac{\mathrm{d} y}{\mathrm{d}t}+ky=0,
\end{equation*}

where $m$ is the mass (or the 'weight'), $c$ is the damping coefficient and $k$ is the *spring constant* which is related to the stiffness of the spring.

**Important**: Notice that since this is a real mechanical system $m$, $c$ and $k$ must all take real values $>0$.

Lets re-write the above as

\begin{equation*}
 \frac{\mathrm{d}^2 y}{\mathrm{d}t^2}+2\alpha\frac{\mathrm{d} y}{\mathrm{d}t}+\omega^2 y=0,
\end{equation*}

where

\begin{equation*}
 \alpha=\frac{c}{2m}, \; \; \; \; \; \; \omega=\sqrt{\frac{k}{m}}.
\end{equation*}

Thus, when we have two distinct real roots, this corresponds to $\alpha^2>\omega^2$ and

\begin{equation*}
 \lambda_{1,2}=-\alpha\pm\sqrt{\alpha^2-\omega^2}.
\end{equation*}

**Important:** Since $\alpha>0$ and real, this means that both values of $\lambda$ **MUST** be negative. This restriction ensures we get sensible results and that our solution doesn't blow up.

**Example**:

Lets consider an example with $m=1$, $c=3$ and $k=2$ with $y(0)=1$ and $\dot{y}(0)=u_0$ which has a solution:

\begin{equation*}
 y = u_0 \left(e^{-t}-e^{-2t}\right)+2e^{-t}+e^{-2t}. 
\end{equation*}

Such oscillators are known as **over-damped**. Lets plot this for a few values of $u_0$:

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

N = 1000
t = np.linspace(0,10,N)

def y_m2r(t, u0):
    return u0*(np.exp(-t)-np.exp(-2*t))+2*np.exp(-t)-np.exp(-2*t)

y1 = y_m2r(t, -1)
y2 = y_m2r(t, 0)
y3 = y_m2r(t, 1)

plt.plot(t, y1, label = "u0=-1")
plt.plot(t, y2, label = "u0=0")
plt.plot(t, y3, label = "u0=1")
plt.xlabel("t")
plt.ylabel("y")
plt.ylim((0, 1.2))
plt.legend()
plt.show()
```

### Two imaginary roots - complex conjugate pair ($4c>b^2$)

From our above analysis, it is clear that complex roots will occur when $b^2-4c<0$ or $4c>b^2$.

The general solution in this case is

\begin{equation*}
 y=Ae^{(p+iq)x}+Be^{(p-iq)x}
\end{equation*}

where
\begin{equation*}
 p=-\frac{b}{2}, \; \; \; \; \; \; q=\frac{\sqrt{4c-b^2}}{2},
\end{equation*}

or

\begin{equation*}
 y=e^{px}\left(Ae^{iqx}+Be^{-iqx}\right)=e^{px}(\hat{A}\cos(qx)+\hat{B}\sin(qx)).
\end{equation*}

**Example**:
    
\begin{equation*}
 \frac{\mathrm{d}^2 y}{\mathrm{d}x^2}+4\frac{\mathrm{d} y}{\mathrm{d}x}+9y=0.
\end{equation*}

Now, $b=4$, $c=9$ and hence:

\begin{equation*}
 \lambda_{1,2}=-2\pm\frac{i\sqrt{4\times 9 - 4^2}}{2}=-2\pm i\sqrt{5}.
\end{equation*}

Solutions therefore take the form:

\begin{equation*}
 y=e^{-2x}(\hat{A}\cos(\sqrt{5}x)+\hat{B}\sin(\sqrt{5}x))
\end{equation*}

When considering mechanical systems with solutions of this form, there are two sub-cases
1. $p=0$
2. $p<0$ (Note we can rule out $p>0$ since this would correspond to the system 'blowing up')

#### case 1: p=0, free-oscillation

This results from oscillatory systems with no damping i.e. $c=0$ and hence $\alpha=0$


\begin{equation*}
 \frac{\mathrm{d}^2 y}{\mathrm{d}t^2}+\omega^2 y=0.
\end{equation*}

**Example**:
    
\begin{equation*}
 \frac{\mathrm{d}^2 y}{\mathrm{d}t^2}+4\pi^2 y=0, \; \; \; \; y(0)=0, \; \; \; \; \frac{\mathrm{d}y}{\mathrm{d}t}(0)=1.
\end{equation*}

Now, $p=0$ and $q=2\pi$ and thus our general solution is

\begin{equation*}
 y(t)=A\cos(2\pi t)+B\sin(2\pi t)
\end{equation*}

Applying our boundary conditions we get:

\begin{equation*}
 A\cos(0)+B\sin(0)=0 \rightarrow A=0,
\end{equation*}

and

\begin{equation*}
 -2\pi A\sin(0)+2\pi B\cos(0)=1 \rightarrow B=\frac{1}{2\pi},
\end{equation*}

and therefore

\begin{equation*}
 y(t)=\frac{1}{2\pi}\sin(2\pi t).
\end{equation*}

Note that this oscillator has an angular velocity $\omega$ of $2\pi$ - that is every 1 unit of time it moves through one full cycle i.e. it's period T=1. A plot:

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

N = 1000
t = np.linspace(0,5,N)

def y_fo(t):
    return 0.5/np.pi*np.sin(2*np.pi*t)

y = y_fo(t)

plt.plot(t, y)
plt.xlabel("t")
plt.ylabel("y")
plt.ylim((-0.2, 0.2))
plt.legend()
plt.show()
```

#### Case 2: Damped oscillator

If we slightly modify the previous example by introducing an $\alpha>0$ but maintaining the relationship $\omega^2 > \alpha^2$ we'll see we now get damped oscillating systems. Lets dive straight into and example:

\begin{equation*}
 \frac{\mathrm{d}^2 y}{\mathrm{d}t^2}+2\sqrt{3}\pi\frac{\mathrm{d}y}{\mathrm{d}t}+4\pi^2 y=0, \; \; \; \; y(0)=0, \; \; \; \; \frac{\mathrm{d}y}{\mathrm{d}t}(0)=u_0.
\end{equation*}

Since $p=-\alpha$ and $q=\sqrt{\omega^2-\alpha^2}$ we obtain

\begin{equation*}
 y(t)=e^{-\sqrt{3}\pi t}(A\cos(\pi t)+B\sin(\pi t)).
\end{equation*}

and applying the boundary conditions:

\begin{equation*}
 y(t)=u_0\frac{e^{-\sqrt{3}\pi t}}{\pi}\sin(\pi t).
\end{equation*}

Notice that the affect of the damping is to reduce the frequency of oscillation (the period is now $T=2$) and to reduce the maximum amplitude over time. A plot of the damped system is:

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

N = 1000
t = np.linspace(0,5,N)

def y_fo(t, u0):
    return u0*np.exp(-np.sqrt(3)*np.pi*t)/np.pi*np.sin(np.pi*t)

y = y_fo(t, 10)

plt.plot(t, y)
plt.xlabel("t")
plt.ylabel("y")
plt.ylim((-0.1, 0.7))
plt.legend()
plt.show()
```

## Non-homogeneous second order linear ODEs

Having examined the homogeneous second order ODE, we now proceed to the
general second order linear ODE with constant coefficients:

\begin{equation}
  y'' + by' + cy  = r(x)
\end{equation}

Similar to first oder non-homogeneous ODEs the solution can again be written as

\begin{equation}
  y = y_h + y_p
\end{equation}

where $y_h$ is the solution to the homogeneous part of the problem and $y_p$ is a particular solution.

We know how to determine $y_h$, so let us examine some methods of determining $y_p$.

### The method of undetermined coefficients and example

In this method we find the particular solution by, based on the form of $r(x)$, guessing a trial solution. This method is great when $r(x)$ takes a relatively simple form, see main notes for further comments.

**MM2 2018 exam q. F2 (i):**

Consider the following ordinary differential equation, which governs the motion of a
mass on spring that is acted on by an external force.

\begin{equation*}
 \frac{\mathrm{d}^2y}{\mathrm{d}t^2}+\omega_0^2y=F_0\sin\omega t.
\end{equation*}

a) Find the general solution of this equation.

b) If the mass starts out with zero velocity, find the particular solution of the above
equation.

**Solution**:

a) The equation is linear and hence we can write the solution in the form

\begin{equation*}
 y=y_h+y_p,
\end{equation*}

where $y_h$ is the solution to the equivalent homogeneous problem and $y_p$ is a specific solution to the non-homogeneous part of the equation. Let us first find $y_h$, which is the solution to

\begin{equation*}
 \frac{\mathrm{d}^2y_h}{\mathrm{d}t^2}+\omega_0^2y_h=0,
\end{equation*}

which has solutions of the form $y\propto e^{\lambda t}$. Looking for solutions of this form we arrive at the characteristic equation

\begin{equation*}
 \lambda^2+\omega_0^2=0,
\end{equation*}

and hence $\lambda$ can take the value $i\omega_0$ or $-i\omega_0$. Again, owing the linearity the solution to homogeneous problem is therefore

\begin{equation*}
 y_h=ae^{i\omega_0 t}+be^{-i\omega_0 t}=c_1\cos\omega_0t+c_2\sin\omega_0t.
\end{equation*}

To find $y_h$ we will use a trial function. Owing to the form of the RHS will will assume

\begin{equation*}
 y_p=A\cos\omega t + B\sin\omega t,
\end{equation*}

and hence

\begin{align*}
 y_p'&=-\omega A\sin\omega t + \omega B\cos\omega t, \\
 y_p''&=-\omega^2 A\cos\omega t - \omega^2 B\sin\omega t.
\end{align*}

Substituting into our original equation we obtain

\begin{equation*}
 -\omega^2 A\cos\omega t - \omega^2 B\sin\omega t + \omega_0^2A\cos\omega t + \omega_0^2B\sin\omega t=F_0\sin\omega t.
\end{equation*}

Equating coefficients of

\begin{align*}
 \cos\omega t: \qquad -\omega^2 A + \omega_0^2A&=0 \; \rightarrow \; A=0, \\
 \sin\omega t: \qquad -\omega^2 B + \omega_0^2B&=F_0 \; \rightarrow \; B=\frac{F_0}{\omega_0^2-\omega^2}.
\end{align*}

and hence the general solution can be written as

\begin{equation*}
 y=c_1\cos\omega_0t+c_2\sin\omega_0t+\frac{F_0}{\omega_0^2-\omega^2}\sin\omega t.
\end{equation*}

b) If $\dot{y}(0)=0$ then

\begin{equation*}
 \dot{y}(0)=-\omega_0c_1\sin(\omega_00)+\omega_0c_2\cos(\omega_00)+\frac{\omega F_0}{\omega_0^2-\omega^2}\cos(\omega 0)=0,
\end{equation*}

and hence

\begin{equation*}
 c_2=-\frac{\omega}{\omega_0}\frac{F_0}{\omega_0^2-\omega^2},
\end{equation*}

yielding

\begin{equation*}
 y=c_1\cos\omega_0t+\frac{F_0}{\omega_0^2-\omega^2}\left(\sin\omega t-\frac{\omega}{\omega_0}\sin\omega_0 t\right),
\end{equation*}

where we note that we are not given a boundary condition that allows us to determine $c_1$.

## Variation of parameters

The method of variation of parameters is a direct generalisation of the
method for first order ODEs. The slight complication is that we now have two
independent solutions from which to work. If the general solution to the
homogeneous problem is:

\begin{equation*}
  y=Ay_1+By_2
\end{equation*}

then we look for a solution of the form:

\begin{equation*}
  y_p=u(x)y_1+v(x)y_2
\end{equation*}

If we substitute this solution into the differential equation, we will have
one equation for two unknown functions. This enables us to impose another
auxiliary equation, which we choose so as to simplify the derivatives:

\begin{equation*}
  u'y_1+v'y_2=0.
\end{equation*}

Differentiating $y_p$, we have:

\begin{equation*}
  \begin{split}
    y'_p&=u'y_1+uy_1'+v'y_2+vy'_2\\
    &=uy'_1+vy'_2
  \end{split}
\end{equation*}

and thence:

\begin{equation*}
  \begin{split}
    y''_p = u'y'_1 + uy_1'' + v'y_2' + vy_2''
  \end{split}
\end{equation*}

Substituting these into the equation yields:

\begin{equation*}
  u(y_1''+by_1'+cy_1)+v(y_2''+by_2'+cy_2)+u'y_1'+v'y_2'=r
\end{equation*}

Since $y_1$ and $y_2$ are solutions to the homogeneous part of the equation,
the expressions in brackets vanish. This leaves:

\begin{equation*}
  u'y_1'+v'y_2'=r
\end{equation*}

By considering this equation and the auxiliary equation together, we can
construct a two-dimensional linear system to solve for $u'$ and $v'$:

\begin{equation*}
  \begin{bmatrix}
    y_1 & y_2 \\
    y_1' & y_2'
  \end{bmatrix}
  \begin{bmatrix}
    u'\\
    v'\\
  \end{bmatrix}
  =
  \begin{bmatrix}
    0\\
    r
  \end{bmatrix}
\end{equation*}

If we write $\textbf{A}$ for the matrix:
\begin{equation*}
  \textbf{A}=  
  \begin{bmatrix}
    y_1 & y_2 \\
    y_1' & y_2'
  \end{bmatrix}
\end{equation*}

then the inverse of $\textbf{A}$ is given by:

\begin{equation*}
  \textbf{A}^{-1}=
  \frac{1}{\det\textbf{A}}
  \begin{bmatrix}
    y_2' & -y_2 \\
    -y_1' & y_1
  \end{bmatrix}
\end{equation*}

where $\det\textbf{A}$ is the **determinant** of $\textbf{A}$, which is given by
$y_1y_2'-y_2y_1'$. Using this inverse matrix we can write:

\begin{equation*}
  \begin{split}
    \begin{bmatrix}
      u'\\
      v'\\
    \end{bmatrix}
    &=
    \textbf{A}^{-1}
    \begin{bmatrix}
      0\\
      r
    \end{bmatrix}\\
    &=
    \frac{1}{\det\textbf{A}}
    \begin{bmatrix}
      y_2' & -y_2 \\
      -y_1' & y_1
    \end{bmatrix}
    \begin{bmatrix}
      0\\
      r
    \end{bmatrix}\\
    &= \frac{1}{\det\textbf{A}}
    \begin{bmatrix}
      -y_2r\\
      y_1r
    \end{bmatrix}
  \end{split}
\end{equation*}

That is to say:

\begin{equation*}
  u'=\frac{-y_2r}{W} \quad\quad v'=\frac{y_1r}{W}
\end{equation*}

The quantity $W\equiv y_1y_2'-y_2y_1'$, the determinant of $\textbf{A}$ is
known as the **Wronskian** after the Polish philosopher and
mathematician Jzef Hoene-Wroski .

Finally we can write expressions for $u$ and $v$:

\begin{equation*}
  u=-\int\frac{y_2r}{W}\mathrm{d}x \quad\quad v=\int\frac{y_1r}{W}\mathrm{d}x
\end{equation*}

In summary, the solution procedure is:
1. Find the homogeneous solutions $y_1$, $y_2$.
2. Calculate $W$, $u$ and $v$.
3. $y_p=uy_1+vy_2$.
4. Form the general solution as $y=y_h+y_p$.

#### Example

Let us again find $y_p$ for the following equation

\begin{equation*}
 \frac{\mathrm{d}^2y}{\mathrm{d}t^2}+\omega_0^2y=F_0\sin\omega t.
\end{equation*}

but this time using variation of parameters. We know that

\begin{equation*}
 y_h=ae^{i\omega_0 t}+be^{-i\omega_0 t}=c_1\cos\omega_0t+c_2\sin\omega_0t.
\end{equation*}

and hence

\begin{align*}
 y_1=e^{i\omega_0 t}, \qquad y_2=e^{-i\omega_0 t}.
\end{align*}

Using the definition above for the Wronskian we have

\begin{equation*}
 W=-2i\omega_0.
\end{equation*}

Thus, since $r(x)=F_0\sin\omega t$ we have

\begin{align*}
 y&=uy_1+vy_2, \\
 u &= -\int\frac{y_2r}{W}\mathrm{dt}=-\int\frac{e^{-i\omega_0t}F_0\sin\omega t}{-2i\omega_0}\mathrm{dt}, \\
 v &= \int\frac{y_1r}{W}\mathrm{dt}=\int\frac{e^{i\omega_0t}F_0\sin\omega t}{-2i\omega_0}\mathrm{dt},
\end{align*}

and hence

\begin{align*}
 u &=-\int\frac{e^{-i\omega t}}{-2i\omega_0}F_0\frac{e^{i\omega t}-e^{-i\omega t}}{2i}\mathrm{d}t \\
   &=-\frac{F_0}{4\omega_0}\int\left(e^{i(\omega-\omega_0)t}-e^{-i(\omega+\omega_0)t}\right)\mathrm{d}t \\
   &=-\frac{F_0}{4\omega_0}\left[\frac{e^{i(\omega-\omega_0)t}}{i(\omega-\omega_0)}+\frac{e^{-i(\omega+\omega_0)t}}{i(\omega+\omega_0)}\right].
\end{align*}

Similarly for $v$:

\begin{align*}
 v &=-\int\frac{e^{i\omega t}}{-2i\omega_0}F_0\frac{e^{i\omega t}-e^{-i\omega t}}{2i}\mathrm{d}t \\
   &=-\frac{F_0}{4\omega_0}\int\left(e^{i(\omega+\omega_0)t}-e^{i(\omega_0-\omega)t}\right)\mathrm{d}t \\
   &=\frac{F_0}{4\omega_0i}\left[\frac{e^{i(\omega+\omega_0)t}}{\omega+\omega_0}-\frac{e^{i(\omega_0-\omega)t}}{\omega_0-\omega}\right].
\end{align*}

Finally, we have that

\begin{align*}
 y_p&=uy_1+vy_2 \\
  &=\frac{F_0}{4\omega_0i}\left[\frac{e^{i\omega t}}{\omega_0+\omega}-\frac{e^{-i\omega t}}{\omega_0-\omega}-\frac{e^{i\omega t}}{\omega-\omega_0}-\frac{e^{-i\omega t}}{\omega_0+\omega}\right] \\
  &=\frac{F_0}{\omega_0^2-\omega^2}\sin\omega t,
\end{align*}

which is, as expected, the same result we found previously.

## Linear ODEs with variable coefficients

The next stage of complexity in the equations is achieved by allowing the
coefficients to be general functions of $x$:

\begin{equation*}
  \frac{\mathrm{d}^2 y}{\mathrm{d}x^2}+b(x)\frac{\mathrm{d}y}{\mathrm{d}x}+c(x)y=r(x)
\end{equation*}

These equations are **much** harder to solve than the constant
coefficient case. A core technique for solution are series approximation
techniques, which we shall examine later. However, some of these equations
can be solved by purely algebraic manipulation.

### The Cauchy or Euler equation

An example of a second order ODE with variable coefficients which is solvable
is the equation variously known as the Cauchy or Euler equation:

\begin{equation*}
  x^2y''+bxy'+cy=0
\end{equation*}

To solve this equation, we first make the substitution $x=e^t$ from which we
have:

\begin{equation*}
  \begin{split}
    \frac{\mathrm{d}y}{\mathrm{d}x}&=\frac{\mathrm{d}y}{\mathrm{d}t}\frac{\mathrm{d}t}{\mathrm{d}x}\\
    &=\frac{\mathrm{d}y}{\mathrm{d}t}e^{-t}\\
    &= \frac{\mathrm{d}y}{\mathrm{d}t}\frac{1}{x}
  \end{split}
\end{equation*}

and:

\begin{equation*}
  \begin{split}
    \frac{\mathrm{d}^2y}{\mathrm{d}x^2}&=\frac{\mathrm{d}}{\mathrm{d}x}\left(\frac{\mathrm{d}y}{\mathrm{d}t}e^{-t}\right)\\
    &=\frac{1}{x}\frac{\mathrm{d}}{\mathrm{d}t}\left(\frac{\mathrm{d}y}{\mathrm{d}t}e^{-t}\right)\\
    &=\frac{1}{x}\left(\frac{\mathrm{d}^2y}{\mathrm{d}t^2}e^{-t}-\frac{\mathrm{d}y}{\mathrm{d}t}e^{-t}\right)\\
    &=\frac{1}{x^2}\left(\frac{\mathrm{d}^2y}{\mathrm{d}t^2}-\frac{\mathrm{d}y}{\mathrm{d}t}\right)
  \end{split}
\end{equation*}

Substituting the derivatives back into the equation, we have:

\begin{equation*}
  \frac{\mathrm{d}^2y}{\mathrm{d}t^2}+(b-1)\frac{\mathrm{d}y}{\mathrm{d}t}+cy=0
\end{equation*}

which has constant coefficients and characteristic equation:

\begin{equation*}
  \lambda^2+(b-1)\lambda+c=0
\end{equation*}

We will examine just a couple of cases. Suppose the roots of the
characteristic equation are real. Then we have the general solution:

\begin{equation*}
  y=Ae^{\lambda_1t}+Be^{\lambda_2t}
\end{equation*}

or

\begin{equation*}
  y=Ax^{\lambda_1}+Bx^{\lambda_2}
\end{equation*}

Alternatively, suppose we have a double root. Then:

\begin{equation*}
  y=(A+Bt)e^{\frac{1-b}{2}t}
\end{equation*}

but, $x=e^t$ gives $x^{\frac{1-b}{2}} = e^{\frac{1-b}{2}t}$ and $t = \ln x$, so:

\begin{equation*}
  y=(A+B \ln x)x^{\frac{1-b}{2}}
\end{equation*}

#### Example

Solve the Cauchy equation:

\begin{equation*}
  y'' - 2\frac{1}{x}y'+2\frac{1}{x^2}y=0.
\end{equation*}

This is the same equation as in problem 3 of the second coursework. Notice that if we multiply the above equation by $x^2$ we get the standard form:

\begin{equation*}
  x^2y'' - 2xy'+2y=0.
\end{equation*}

This yields the characteristic equation

\begin{equation*}
  \lambda^2-3\lambda+2=0,
\end{equation*}

giving $\lambda_1=1$, $\lambda_2=2$ and hence

\begin{equation*}
  y=Ax+Bx^2.
\end{equation*}

# Power series

## Power series expansions

A general power series expansion can take the form:

\begin{equation*}
  f(x) = a_0 + a_1x + a_2x^2 + \ldots + a_n x^n + \ldots = \sum_{k=0}^{\infty} a_k x^k.
\end{equation*}

We can also write this as

\begin{equation*}
  f(x)=b_0+b_1(x-x_0)+b_2(x-x_0)^2+\ldots+b_n(x-x_0)^n+\ldots = \sum_{k=0}^{\infty} b_k (x-x_0)^k,
\end{equation*}

which is common if expanding a function around $x_0$. From this we can derive the following form of the Taylor expansion

\begin{equation*}
  \begin{split}
    f(x)&=\sum_{k=0}^\infty \frac{f^{(k)}(x_0)}{k!}(x-x_0)^k.
  \end{split}
\end{equation*}

Manipulating this we can also write down the following expansions:

\begin{equation*}
  f(x+x_0)=\sum_{k=0}^\infty \frac{f^{(k)}(x_0)}{k!}x^k
\end{equation*}

\begin{equation*}
  f(x)=\sum_{k=0}^\infty \frac{f^{(k)}(0)}{k!}x^k
\end{equation*}

This final form of the Taylor series is referred to as a MacLaurin series. The radius of convergence of the above expansions is defined by

\begin{equation*}
  R\equiv \lim_{n\rightarrow\infty}\left|\frac{a_{n}}{a_{n+1}}\right|,
\end{equation*}

that is, what does the ratio of the *next*, compared to the *current*, coefficient tend to as $n\rightarrow\infty$.

**Example: MM2 2018 Exam F2(ii)**

Find the first three, non-zero terms of the Taylor series of $\ln(1+x)$. You may assume
that the Taylor series of a function $f(x+x_0)$ is given by,

\begin{equation*}
  f(x+x_0)=\sum_{k} \frac{f^{(k)}(x_0)}{k!}x^k.
\end{equation*}

**Solution:**

Evaluating the first few derivatives of our function we have:

\begin{align*}
 f'(x)&=\frac{1}{1+x}, \\
 f''(x)&=\frac{-1}{(1+x)^2}, \\
 f'''(x)&=\frac{2}{(1+x)^3}
\end{align*}

Note that $x_0$ is not known and $\ln(1+x_0)$ may be zero is $x_0=0$ and hence we will write

\begin{equation*}
 f(x+x_0)=\ln(1+x_0)+\frac{1}{1+x_0}x-\frac{1}{2(1+x_0)^2}x^2+\frac{1}{3(1+x_0)^3}x^3+\ldots,
\end{equation*}

or

\begin{equation*}
 f(x+x_0)=\ln(1+x_0)+\sum_{k=1}^{\infty}\frac{(-1)^{k+1}}{k(1+x_0)^k}x^k.
\end{equation*}

Although not part of the exam question, for completeness lets consider the radius of convergence. This is given by:

\begin{equation*}
  R = \lim_{n\rightarrow\infty}\left|(1+x_0)\frac{n+1}{n}\right|=1+x_0.
\end{equation*}

## Power series solutions to ODEs

A common method for solving linear PDEs with non-constant coefficients (which aren't in a common form) is to look for power series solutions. To understand this better and to analyze the types of expansions we can use we can extend and study the differential equations in the complex plane - **however** this is not something we studied during the course and hence not examinable. That is, you will only be required to solve simple examples such as that presented below using Frobenius' method.

## Frobenius' method

When using Frobenius' method we look for solutions of the form

\begin{equation*}
 y=x^m\sum_ka_kx^k,
\end{equation*}

**Example:**

Solve

\begin{equation*}
 xy''-2y'+9x^5y=0.
\end{equation*}

Differentiating our trial solution we have

\begin{align*}
 y'&=\sum_k(m+k)a_kx^{k+m-1}, \\
 y''&=\sum_k(m+k-1)(m+k)a_kx^{k+m-2},
\end{align*}

and substituting in we arrive at

\begin{align*}
 \sum_k(m+k-1)(m+k)a_kx^{k+m-1}-2\sum_k(m+k)a_kx^{k+m-1}+9\sum_ka_kx^{m+k+5}=0.
\end{align*}

or

\begin{align*}
 \sum_k(m+k-3)(m+k)a_kx^{k+m-1}+9\sum_ka_kx^{m+k+5}=0.
\end{align*}

Comparing coefficients at $x^{m-1}$ we see that

\begin{equation*}
 m(m-3)=0,
\end{equation*}

and hence $m=0$ or $m=3$. First, to find the general form of the expansion when $m=0$ let us analyse

\begin{equation*}
 \sum_k(k-3)ka_kx^{k-1}+9\sum_ka_kx^{k+5}=0.
\end{equation*}

This tells us that

\begin{align*}
 6k(6k-3)a_{6k}+9a_{6k-6}&=0, \\
 a_{6k}=-\frac{9a_{6k-6}}{6k(6k-3)}=-\frac{1}{2k(2k-1)}a_{6k-6}&=\frac{(-1)^k}{(2k)!}a_0.
\end{align*}

Hence

\begin{equation*}
 y_1=a_0\sum_{k=0}^{\infty}\frac{(-1)^k}{(2k)!}x^{6n}=a_0\cos(x^3).
\end{equation*}

Now, when $m=3$

\begin{equation*}
 \sum_kk(k+3)a_kx^{k+2}+9\sum_ka_kx^{k+8}=0.
\end{equation*}

Notice there is again a 'power of six' relationship between indices. This give us the relationship

\begin{align*}
 6k(6k+3)a_{6k}+9a_{6k-6}&=0, \\
 a_{6k}=-\frac{9a_{6k-6}}{6k(6k+3)}=-\frac{1}{2k(2k+1)}a_{6k-6}&=\frac{(-1)^k}{(2k+1)!}a_0.
\end{align*}

and hence

\begin{equation*}
 y_2=a_0x^3\sum_{k=0}^{\infty}\frac{(-1)^k}{(2k+1)!}x^{6n}=a_0\sum_{k=0}^{\infty}\frac{(-1)^k}{(2k+1)!}x^{3(2k+1)}=a_0\sin(x^3).
\end{equation*}

The solution is thus

\begin{equation*}
 y=A\sin(x^3)+B\cos(x^3).
\end{equation*}

# Fourier series

We've just reviewed expanding functions as polynomial series. When looking for series expansions of periodic functions it's useful to switch to a series that makes use of periodic basis functions - this is what the Fourier series is. Lets review Fourier series by considering a past exam question.

**Example: MM2 2019 Q3:**

The coefficients of the Fourier series are defined by,

\begin{align*}
 a_0&=\frac{1}{\pi}\int_{-\pi}^{\pi}f(x)\mathrm{d}x, \\
 a_n&=\frac{1}{\pi}\int_{-\pi}^{\pi}f(x)\cos(nx)\mathrm{d}x, \\
 b_n&=\frac{1}{\pi}\int_{-\pi}^{\pi}f(x)\sin(nx)\mathrm{d}x,
\end{align*}

(a) Sketch the following function, in an interval $-\pi < x < \pi$:

\begin{equation*}
 f(x)=\begin{cases}
    -x&, \qquad -\pi<x<0 \\
    x&,  \qquad 0<x<\pi
\end{cases}
\end{equation*}

(b) Express the function $f(x)$, in an interval $-\pi < x < \pi$, as a Fourier series.

**Solution:**

(a) First lets plot the function, here we'll use a little bit of python:

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

N = 1000
x = np.linspace(-np.pi,np.pi,N)

def fp(x):
    return np.abs(x)

plt.plot(x, fp(x))
plt.xlabel("$x$")
plt.ylabel("$f(x)$")
plt.ylim((-0.1, 1.1*np.pi))
plt.show()
```

***Note*** that in an exam, since you'll be sketching any required plots on paper be sure to be very clear with value labeling - i.e. be sure to clearly label the values of $f(x)$ at the origin and at the edge of periodic windows (and any other important, e.g. turning, points if present).  

(b)Our Fourier series will be of the form

\begin{equation*}
 f(x)=\frac{a_0}{2}\sum_{n=1}^{\infty}\left[a_n\cos(nx)+b_n\sin(nx)\right].
\end{equation*}

We will now evaluate the coefficients $a_0$, $a_n$ and $b_n$ utilizing the provided formulas or otherwise.

We notice that $f(x)=f(-x)$, that is $x=-(-x)$, and the function is therefore even.
Since the function is even, the series must contain only even basis functions. Therefor $b_n=0 \; \forall \; n$. Further, since $a_0/2$ represents the average of the series over the periodic window, this is clearly $\pi/2$, that is

\begin{equation*}
 \frac{a_0}{2}=\frac{\pi}{2}.
\end{equation*}

With the above in place, we wouldn't need to actually compute $a_0$ in the exam, but for completeness lets just do it here:

\begin{align*}
 a_0&=\frac{1}{\pi}\int_{-\pi}^{\pi}f(x)\mathrm{d}x, \\
    &=\frac{1}{\pi}\int_{-\pi}^{0}-x\mathrm{d}x+\frac{1}{\pi}\int_{0}^{\pi}x\mathrm{d}x \\
    &=\frac{1}{\pi}\left(\left[-\frac{x^2}{2}\right]_{-\pi}^{0}+\left[\frac{x^2}{2}\right]_{0}^{\pi}\right) \\
    &=\pi.
\end{align*}

Let us now compute $a_n$.

\begin{align*}
 a_n&=\frac{1}{\pi}\int_{-\pi}^{\pi}f(x)\cos(nx)\mathrm{d}x \\
    &=\frac{1}{\pi}\int_{-\pi}^{0}-x\cos(nx)\mathrm{d}x+\frac{1}{\pi}\int_{0}^{\pi}x\cos(nx)\mathrm{d}x
\end{align*}

Let I = $\int_{a}^{b} x\cos(nx)\mathrm{d}x$. Applying integration by parts we have

\begin{align*}
 I=&\left[\frac{x\sin(nx)}{n}\right]_{a}^{b}-\int_{a}^{b}\frac{\sin{nx}}{n}\mathrm{d}x \\
  =&\left[\frac{x\sin(nx)}{n}+\frac{\cos(nx)}{n^2}\right]_{a}^{b}.
\end{align*}

Utilizing this result in our expression for $a_n$ we arrive at

\begin{align*}
 a_n&=-\frac{1}{\pi}\left[\frac{x\sin(nx)}{n}+\frac{\cos(nx)}{n^2}\right]_{-\pi}^{0}+\frac{1}{\pi}\left[\frac{x\sin(nx)}{n}+\frac{\cos(nx)}{n^2}\right]_{0}^{\pi} \\
   &=-\frac{1}{\pi}\left[\frac{1}{r^2}-\frac{\cos n\pi}{n^2}\right]+\frac{1}{\pi}\left[\frac{\cos n\pi}{n^2}-\frac{1}{n^2}\right] \\
   &=\frac{2}{\pi n^2}\left[\cos n\pi - 1\right] \\
   &=\begin{cases}
    -\frac{4}{\pi n^2}, \qquad &n \;\mathrm{odd} \\
    0, \qquad &n \;\mathrm{even}
    \end{cases}
\end{align*}

Although we know $b_n=0 \; \forall \; n$, since we're revising lets just go through this calculation and convince ourselves this is indeed the case. We have

\begin{align*}
 b_n&=\frac{1}{\pi}\int_{-\pi}^{\pi}f(x)\sin(nx)\mathrm{d}x \\
    &=\frac{1}{\pi}\int_{-\pi}^{0}-x\sin(nx)\mathrm{d}x+\frac{1}{\pi}\int_{0}^{\pi}x\cos(nx)\mathrm{d}x \\
    &=\frac{1}{\pi}\left(\left[\frac{x\cos(nx)}{n}\right]_{-\pi}^{0}
    -\int_{-\pi}^{0}\frac{\cos(nx)}{n}\mathrm{d}x\right.+
    \left.\left[\frac{-x\cos(nx)}{n}\right]_{0}^{\pi}
    +\int_{0}^{\pi}\frac{\cos(nx)}{n}\mathrm{d}x\right) \\
    &=\frac{1}{\pi}\left(\pi\cos(-n\pi)-\pi\cos(n\pi)\right) \\
    &=0
\end{align*}
   
Finally, we have

\begin{equation*}
 f(x)=\frac{\pi}{2}-\frac{4}{\pi}\sum_{s=0}^{\infty}\frac{\cos((2s+1)x)}{(2s+1)^2}.
\end{equation*}

Equivalently, we could have written this as

\begin{equation*}
 f(x)=\frac{\pi}{2}-\frac{4}{\pi}\sum_{s=1}^{\infty}\frac{\cos((2s-1)x)}{(2s-1)^2}.
\end{equation*}

Since we're not currently in an exam, it's good practice to check our result with a bit of python:

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

N = 1000
x = np.linspace(-np.pi,np.pi,N)

def eval_fmx(x,N):
    f = np.pi/2
    for i in range(0,N+1):
        f += -4/np.pi*np.cos((2*i+1)*x)/(2*i+1)**2
    return f

y1 = eval_fmx(x,1)
y2 = eval_fmx(x,2)
y3 = eval_fmx(x,5)
y4 = eval_fmx(x,10)
y5 = eval_fmx(x,100)

fig=plt.figure(figsize=(12,8), dpi= 100, facecolor='w', edgecolor='k')
plt.plot(x, y1, label = "$N=1$")
plt.plot(x, y2, label = "$N=2$")
plt.plot(x, y3, label = "$N=5$")
plt.plot(x, y4, label = "$N=10$")
plt.plot(x, y5, label = "$N=100$")
plt.xlabel("x")
plt.ylabel("y")
plt.ylim((-0.1,1.1*np.pi))
plt.legend()
plt.show()
```

# Multiple integrals

In the final content based lecture we covered multiple integrals. In this course, we have concerned ourselves more with the application than theory of these methods and lets therefore dive straight into an example:

**Question:**

Consider a sphere with radius $r$, azimuthal angle $0\leq\phi\leq2\pi$ and inclination angle $0\leq\theta\leq\pi$ where $\theta=0$ points towards the north pole. Find the volume of the spherical section with the following bounds:

\begin{align*}
 1 &\leq r \leq 2, \\
 \frac{\pi}{4} &\leq \phi \leq \frac{3\pi}{4}, \\
 \frac{\pi}{4} &\leq \theta \leq \frac{2\pi}{3}.
\end{align*}

Then compute the integral of the function $f(x,y,z)=x^2+y^2+z^2$ over the same segment.

**Solution:**
    
In the given coordinate system,

\begin{align*}
 x&=r\cos\phi\sin\theta, \\
 y&=r\sin\phi\sin\theta, \\
 z&=r\cos\theta
\end{align*}

And the integral of a spherical element is given by

\begin{equation*}
 I=\iiint f(r,\phi,\theta)r^2\sin\theta\mathrm{d}r\mathrm{d}\theta\mathrm{d}\phi.
\end{equation*}

Since we're computing the volume, $f(r,\phi,\theta)=1$ and hence our integral becomes

\begin{align*}
 V&=\int_{1}^{2}\int_{\frac{\pi}{4}}^{\frac{2\pi}{3}}\int_{\frac{\pi}{4}}^{\frac{3\pi}{4}}r^2\sin\theta\mathrm{d}\phi\mathrm{d}\theta\mathrm{d}r \\
  &=\frac{\pi}{2}\int_{\frac{\pi}{4}}^{\frac{2\pi}{3}}\int_{1}^{2}r^2\sin\theta\mathrm{d}r\mathrm{d}\theta \\
  &=-\frac{\pi}{2}\int_{1}^{2}[\cos\theta]_{\frac{\pi}{4}}^{\frac{2\pi}{3}}r^2\mathrm{d}r \\
  &=\frac{\pi(1+\sqrt{2})}{4}\left[\frac{r^3}{3}\right]_{1}^{2} \\
  &=\frac{7(1+\sqrt{2})\pi}{12}.
\end{align*}

Now, we repeat this but with $f(x,y,z)=x^2+y^2+z^2$. Converting to polar form we can show that $f(r,\phi,\theta)=r^2$. Thus:

\begin{align*}
 I&=\frac{\pi(1+\sqrt{2})}{4}\left[\frac{r^5}{5}\right]_{1}^{2} \\
  &=\frac{31(1+\sqrt{2})\pi}{20}.
\end{align*}
