# Lecture 2 - Complex numbers II, Ordinary Differential Equations I



One of the important things we learned in lecture one was the identity

\begin{equation*}
  e^{i\theta}=\cos\theta+i\sin\theta,
\end{equation*}

Allowing us to relate general complex numbers in various forms:

\begin{equation*}
  z=x+iy=r(\cos\theta+i\sin\theta)=re^{i\theta},
\end{equation*}

where

\begin{equation*}
  r=|z|=\sqrt{z\bar{z}}, \qquad \text{and} \;\; \bar{z}=x-iy.
\end{equation*}

Today's we'll delve a little deeper into some of the properties of complex numbers and some of the relationships we can derive through using these properties.

## de Moivreâ€™s identity

Further, this exponential form provides a mechanism for raising a complex number to a power:

\begin{gather*}
  z=re^{i\theta}\\
  z^n=\left(re^{i\theta}\right)^n\\
  z^n=r^ne^{in\theta}
\end{gather*}

Further insight can be gained by expressing both sides in terms of
$\cos\theta+i\sin\theta$:

\begin{gather*}
  r^n(\cos\theta+i\sin\theta)^n=r^n(\cos n\theta+i\sin n\theta)\\
  (\cos\theta+i\sin\theta)^n=\cos n\theta+i\sin n\theta
\end{gather*}

This formula is known as **de Moivre's formula** after the French Hugenot
mathematician and London resident Abraham de Moivre.

### Example

Express:

\begin{equation*}
  \cos3\theta
\end{equation*}

in terms of $\cos^n\theta$ and $\sin^n\theta$.

Using de Moivre's formula:

\begin{gather*}
  (\cos\theta+i\sin\theta)^3=\cos3\theta+i\sin3\theta\\
  \cos^3\theta + 3i\cos^2\theta\sin\theta  - 3\sin^2\theta\cos\theta -
  i\sin^3\theta = \cos3\theta+i\sin3\theta \\
  \text{Gathering terms:} \\
  i\sin3\theta=3i\cos^2\theta-i\sin^3\theta\\
  \cos3\theta=\cos^3\theta-3\sin^2\theta\cos\theta\\
  \cos3\theta=\cos^3\theta-3(1-\cos^2\theta)\cos\theta\\
  \cos3\theta=4\cos^3\theta-3\cos\theta\\
\end{gather*}

## Solving complex equations

We're familiar with solving sets of equations containing real unknowns, e.g.

\begin{align*}
  x + y = 0, \\
  2x - y = 3.
\end{align*}

With a little bit of algebra we can then show $x=1$, $y=-1$ is a solution to this set of equations.

Now, consider the complex equation

\begin{equation*}
  \bar{z}+\frac{i}{2}(z+3\bar{z})=3i.
\end{equation*}

We can now ask 'which value of $z$' satisfies this equation? (Note that here we only have a single equation here).
To solve this, lets split it up into its real and imaginary components:

\begin{align*}
  x-yi+\frac{i}{2}(x+iy+3x-3iy)=3i, \\
  x+y+i(2x-y)=3i.
\end{align*}

Taking real and imaginary parts we have

\begin{align*}
  x + y = 0, \\
  2x - y = 3.
\end{align*}

This is just the set of real equations we looked at above! Hence, the solution is $z=1-i$. We'll look at more examples of complex equations, including an example from last years exam in the problem session later.

## Complex numbers with modulus 1

Since for such numbers $|z|=r=1$, complex numbers with a modulus of 1 can be expressed
$z=\cos\theta+i\sin\theta$. Now consider:

\begin{align*}
  z^n+\frac{1}{z^n}&=
  \cos n\theta+i\sin n\theta+\frac{1}{\cos n\theta+i\sin n\theta}\\
  &=\cos n\theta+i\sin n\theta+\cos n\theta-i\sin n\theta\\
  &=2\cos n\theta
\end{align*}
Similarly:
\begin{align*}
  z^n-\frac{1}{z^n}&=2i\sin n\theta
\end{align*}

### Example

Express $\cos^3\theta$ in terms of $\cos n\theta$. Using the above
expressions, if $z=\cos\theta+i\sin\theta$ then:

\begin{align*}
  2\cos\theta&=\left(z+\frac{1}{z}\right)\\
  2^3\cos^3\theta&=\left(z+\frac{1}{z}\right)^3\\
  &=z^3+3z+\frac{3}{z}+\frac{1}{z^3}\\
  &=\left(z^3+\frac{1}{z^3}\right)+3\left(z+\frac{1}{z}\right)\\
  &=2\cos3\theta+3\times2\cos\theta
\end{align*}

And hence:

\begin{align*}
  \cos^3\theta=\frac{1}{4}\cos3\theta+\frac{3}{4}\cos\theta
\end{align*}

### Verification

Using

\begin{equation*}
 \left(re^{i\theta}\right)^n=r^n(\cos(n\theta)+i\sin(n\theta))
\end{equation*}

We can show that (you'll prove this in the problem class later)

\begin{equation*}
 \cos^5(\theta)=\frac{1}{16}\cos(5\theta)+\frac{5}{16}\cos(3\theta)+\frac{5}{8}\cos(\theta).
\end{equation*}

Lets check this is correct:

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

N = 1000
theta = np.linspace(-2*np.pi,2*np.pi,N)

y1 = np.cos(theta)**5
y2 = 1/16*np.cos(5*theta)+5/16*np.cos(3*theta)+5/8*np.cos(theta)

plt.plot(theta, y1, label = "cos(x)**5")
plt.plot(theta, y2, label = "cos(nx) form")
plt.xlabel("theta")
plt.ylabel("f(theta)")
plt.title('Comparison of our two forms')
plt.legend()
plt.show()
```

## Roots of complex numbers

The $n$-th root of a complex number $z$ may be written:

\begin{equation*}
  w=z^{1/n} \qquad\textrm{or}\qquad w=\sqrt[n]{z}
\end{equation*}

We will write $z$ and $w$ as:

\begin{equation*}
  z=r(\cos\theta+i\sin\theta) \qquad\textrm{and}\qquad w=R(\cos\phi+i\sin\phi)
\end{equation*}

Now we must solve:

\begin{gather*}
  w^n=z\\
  R^n(\cos\phi+i\sin\phi)^n=r(\cos\theta+i\sin\theta)\\
  R^n(\cos n\phi+i\sin n\phi)=r(\cos\theta+i\sin\theta)
\end{gather*}

Equating the moduli we have:

\begin{gather*}
  r=R^n\\
  R=r^{1/n}
\end{gather*}

where R is a positive root as it represents the magnitude. Similarly, equating the arguments and remembering that angles separated by
whole multiples of $2\pi$ are equal, we have:

\begin{gather*}
  n\phi=\theta+2k\pi\\
  \phi=\frac{\theta}{n}+\frac{2k\pi}{n}
\end{gather*}

This will produce distinct values of $\phi$ for $k=\ldots,-1,0,1\ldots,n-1$ (that is, $k\in\mathbb{Z}$).

So finally:

\begin{equation*}
  w=r^{1/n}\left(\cos\left(\frac{\theta}{n}+\frac{2k\pi}{n}\right) 
    + i\sin\left(\frac{\theta}{n}+\frac{2k\pi}{n}\right)\right)
\end{equation*}

### Example: the cube roots of unity

Find:

\begin{equation*}
  w=\sqrt[3]{1}
\end{equation*}

Representing $1$ in polar coordinates, we have:

\begin{equation*}
  r=1\qquad \theta=0
\end{equation*}

so that:

\begin{equation*}
  w=1^{1/3}\left(\cos\left(\frac{2k\pi}{3}\right)
    +i\sin\left(\frac{2k\pi}{3}\right)\right)\qquad k=0,1,2
\end{equation*}

We can now evaluate the individual roots:

\begin{align*}
  \text{For $k=0$:} \qquad w_0&=1(1+0i)=1, \\
  \text{For $k=1$:} \qquad w_1&=1\left(\cos\left(\frac{2\pi}{3}\right)+i\sin\left(\frac{2\pi}{3}\right)\right)
  =-\frac{1}{2}+i\frac{\sqrt{3}}{2}, \\
  \text{For $k=2$:} \qquad
  w_2&=1\left(\cos\left(\frac{4\pi}{3}\right)+i\sin\left(\frac{4\pi}{3}\right)\right)
  =-\frac{1}{2}-i\frac{\sqrt{3}}{2}, \\
  \text{For $k=3$:} \qquad
  w_2&=1\left(\cos2\pi+i\sin2\pi\right)
  =1, \qquad
  \text{Note that we recover $w_0$.}
\end{align*}

It is illustrative to plot these roots:

```python
def argand(z, xl=-1.2, xr=1.2, yl=-1.2, yr=1.2):

    rc = {"xtick.direction" : "inout", "ytick.direction" : "inout",
          "xtick.major.size" : 5, "ytick.major.size" : 5,}
    with plt.rc_context(rc):
        fig, ax = plt.subplots(figsize=(6,4), dpi= 120, facecolor='w', edgecolor='k')
        ax.scatter(z.real,z.imag)

        ax.spines['left'].set_position('zero')
        ax.spines['right'].set_visible(False)
        ax.spines['bottom'].set_position('zero')
        ax.spines['top'].set_visible(False)
        ax.xaxis.set_ticks_position('bottom')
        ax.yaxis.set_ticks_position('left')
        ax.set_xlim([xl, xr])
        ax.set_ylim([yl, yr])
        ax.set_aspect('equal')
        ax.set_xlabel('Re(z)', loc='right')
        ax.set_ylabel('Im(z)', loc='top')

        # make arrows
        ax.plot((1), (0), ls="", marker=">", ms=10, color="k",
               transform=ax.get_yaxis_transform(), clip_on=False)
        ax.plot((0), (1), ls="", marker="^", ms=10, color="k",
                transform=ax.get_xaxis_transform(), clip_on=False)

        plt.show()
```

```python
def roots(N,M):
    roots = []
    for i in N:
        roots.append(np.exp(2j*i*np.pi/M))
    return np.array(roots)

N = list(range(-10, 10))
         
z3 = roots(N,3)
    
argand(z3)
```

This allows us to make a number of observations:

- there are three roots.
- one is the usual real cube root.
- two are complex.
- the roots are equally spaced around a circle of radius 1 and the angle between adjacent roots is $2\pi/3$.

These conclusions actually generalise to $z^{1/n}$:
    
- there are $n$ distinct roots.
- the roots are equally spaced with the angle between adjacent roots $2\pi/n$.
- the roots lie on the circumference of a circle with radius $r^{1/n}$ where $r$ is the modulus of $z$.
- complex roots of **real** numbers occur in conjugate pairs: if $z'$ is a root then $\bar{z}'$ is also a root. This is not true for the roots of other complex numbers.

### General root's of unity

We've seen that unity can be expressed in the form $1=e^{i2n\pi}$ for any integer $n$.

Hence, we see that $1^{1/2}=e^{i2n\pi\times\frac{1}{2}}=e^{in\pi}$, $1^{1/3}=e^{i2n\pi/3}$ and so forth.

The unique $k$-th roots of unity will correspond to the 'amount' of complex numbers of the form $e^{i2n\pi/m}$ that fit into *any* $2\pi$ window. 

```python
z4 = roots(N,8)
    
argand(z4)
```

The roots of any other complex number will be closely related to the roots of unity. For a complex number of the form $z=re^{i\theta}$, the $k$-th roots will be given by $r^{1/k}e^{i\theta/k}e^{i2n\pi/k}$. Notice the relation between the roots of $1^{1/8}$ (shown above) and $2e^{i\pi/4}$ (shown below):

```python
def rootsr(r,theta,N,M):
    roots = []
    for i in N:
        roots.append(r**(1/M)*np.exp(1j*theta/M)*np.exp(2j*i*np.pi/M))
    return np.array(roots)

z5 = rootsr(2,0.25*np.pi,N,8)
    
argand(z5)
```

## Ordinary differential equations: types and general considerations

First order ordinary differential equations were covered in Mathematical
Methods I. In this course we will extend this to second order.

Mathematical equations involving derivatives are known as differential
equations. There is a further classification depending on the types of
derivatives:

- Ordinary differential equations (ODEs) involve full derivatives. For example:

  \begin{equation*}
    \frac{\mathrm{d}^2f}{\mathrm{d}x^2}+2x^2\frac{\mathrm{d}f}{\mathrm{d}x}+f=\cos x,
  \end{equation*}
  
- Partial differential equations (PDEs) involve partial derivatives. For example:

  \begin{equation*}
    \frac{\partial f}{\partial x}+\frac{\partial f}{\partial y}=4xy,
  \end{equation*}
  
  where $f=f(x,y)$.

This course will cover only ordinary differential equations. Partial differential equations play a crucial role in the physical sciences because they describe such a large range of phenomenon. However, before we can solve them we need to know how to solve a large range of ODEs.

### Some definitions

The **order** of a differential equation is the order of the highest derivative that it contains:

\begin{eqnarray*}
  \frac{\mathrm{d}y}{\mathrm{d}x}-3x=0 \qquad & \text{1st order}\\
  \frac{\mathrm{d}^2y}{\mathrm{d}x^2}+4\frac{\mathrm{d}y}{\mathrm{d}x}+y=\cos x \qquad &\textrm{2nd order}
\end{eqnarray*}

An equation is said to be **homogeneous** if all terms contain the functional:

\begin{equation*}
 \frac{\mathrm{d}^2y}{\mathrm{d}x^2}+4\frac{\mathrm{d}y}{\mathrm{d}x}+y=0.
\end{equation*}

An equation is said to be **inhomogeneous** if this is not the case

\begin{equation*}
 \frac{\mathrm{d}^2y}{\mathrm{d}x^2}+4\frac{\mathrm{d}y}{\mathrm{d}x}+y=x^2+1.
\end{equation*}

An equation is said to be **linear** if no terms contain the a derivative (including the zeroth derivative) of the functional multiplied some other derivative of the functional:

\begin{equation*}
 x^2\frac{\mathrm{d}^2y}{\mathrm{d}x^2}+4\frac{\mathrm{d}y}{\mathrm{d}x}+y=\cos x, \qquad \text{Linear},
\end{equation*}

\begin{equation*}
 y\frac{\mathrm{d}^2y}{\mathrm{d}x^2}+4\frac{\mathrm{d}y}{\mathrm{d}x}+y=\cos x, \qquad \text{Nonlinear},
\end{equation*}

\begin{equation*}
 2\frac{\mathrm{d}^2y}{\mathrm{d}x^2}+4\frac{\mathrm{d}y}{\mathrm{d}x}+y^2=\cos x, \qquad \text{Nonlinear}.
\end{equation*}

**The linearity of an equation is one of it's most defining and therefore important characteristics. Analytic solutions of linear are far, far more common than those of nonlinear equations.**

Also importantly, solutions of linear ODEs can be added together to create new solutions. That is, if $f_1(x)$ and $f_2(x)$ are solutions to some ODE, then $f(x)=f_1(x)+f_2(x)$ is also a solution. This is not true for nonlinear equations (see exercise sheet later).

### Simple pendulum

Now, lets consider the classic problem of the simple pendulum.

<img src="figures/sp.png" width=450 height=4500 />

The motion is described by Newton's second law: **Force = Mass x Acceleration**.

We'll make some simplifying assumptions:

- The mass (at the end of the string) can be treated as a point mass.
- This mass is much heavier than the string.
- The string does not stretch.
- The system is frictionless
- (and so on).

Motion is restricted along the arc of the string. Hence we can describe the motion in terms of the strings angle to the vertical, $\theta$.

- The force along this arc is $mg\sin\theta$.
- The velocity of the mass along this arc is $l\frac{\mathrm{d}\theta}{\mathrm{d}t}$ and the acceleration is then the derivative of this with respect to time.

Using Newton's second law we hence obtain:

\begin{equation*}
 ml\frac{\mathrm{d}^2\theta}{\mathrm{d}t^2}=-mg\sin\theta,
\end{equation*}

or

\begin{equation*}
 \frac{\mathrm{d}^2\theta}{\mathrm{d}t^2}+\frac{g}{l}\sin\theta=0.
\end{equation*}

The equation is homogeneous, but oh dear...







...

...

# IT'S NONLINEAR....

owing to the $\sin\theta$ term.

Solutions are in fact related to 'elliptic integrals' (beyond the scope of this course). However, we can make progress through utilizing the important method of *linearisation*.

The Taylor-Maclaurin expansion of $\sin\theta$ is

\begin{equation*}
 \sin\theta\approx\theta-\frac{\theta^3}{3!}+\ldots.
\end{equation*}

Hence, is we assume $\theta$ is small and make the approximation

\begin{equation*}
 \sin\theta\approx\theta,
\end{equation*}

we will then have a linear equation (the unknown is not multiplied by itself or any of its derivatives) which we can make progress with:

\begin{equation*}
 \frac{\mathrm{d}^2\theta}{\mathrm{d}t^2}-\frac{g}{l}\theta=\left(\frac{\mathrm{d}^2}{\mathrm{d}t^2}-\frac{g}{l}\right)\theta=0.
\end{equation*}

To solve this, lets first factorize the differential operator:

\begin{equation*}
 \left(\frac{\mathrm{d}}{\mathrm{d}t}+i\sqrt{\frac{g}{l}}\right)\left(\frac{\mathrm{d}}{\mathrm{d}t}-i\sqrt{\frac{g}{l}}\right)\theta=0.
\end{equation*}

Thus, for this we equation to be satisfied, we need

\begin{equation*}
 \frac{\mathrm{d}\theta}{\mathrm{d}t}+i\sqrt{\frac{g}{l}}\theta=0,
\end{equation*}

or

\begin{equation*}
 \frac{\mathrm{d}\theta}{\mathrm{d}t}-i\sqrt{\frac{g}{l}}\theta=0.
\end{equation*}

These are two first order ODEs which we know how to solve via separation of variables. The solutions are
\begin{align*}
 \theta_1(t)&=c_1e^{-i\sqrt{\frac{g}{l}}t}, \\
 \theta_2(t)&=c_2e^{i\sqrt{\frac{g}{l}}t}.
\end{align*}

Further, this equation is **linear** and hence we know

\begin{equation*}
 \theta=\theta_1+\theta_2=c_1e^{-i\sqrt{\frac{g}{l}}t}+c_2e^{i\sqrt{\frac{g}{l}}t},
\end{equation*}

is also a solution! This is in fact the general solution to this ODE. Through using

\begin{equation*}
 e^{i\phi}=cos\phi+i\sin{\phi},
\end{equation*}

this can be re-written (algebra left to problem sheet) as

\begin{equation*}
 \theta = A\cos\left(\sqrt{\frac{g}{l}}t\right)+B\sin\left(\sqrt{\frac{g}{l}}t\right).
\end{equation*}

Note that the constants $A$ and $B$, although related are different from $c_1$, $c_2$.

Finally, if we wish to determine the constants $A$ and $B$ we will need some initial conditions. Given that we have two unknown constants, we'll need two conditions.

Lets say that at time $t=0$, $\theta=\frac{\pi}{6}$ and $\frac{\mathrm{d}\theta}{\mathrm{d}t}=\dot{\theta}=0$.

The first condition implies

\begin{equation*}
 \theta(0)=\frac{\pi}{6}=A.
\end{equation*}

The second condition implies

\begin{equation*}
 \dot{\theta}=0=-\frac{\pi}{6}\sqrt{\frac{g}{l}}\sin\left(\sqrt{\frac{g}{l}}\times 0\right)+B\sqrt{\frac{g}{l}}\cos\left(\sqrt{\frac{g}{l}}\times 0\right) \; \rightarrow \; B=0.
\end{equation*}

Hence, the **specific solution** given these initial conditions is

\begin{equation*}
 \theta=\frac{\pi}{6}\cos\left(\sqrt{\frac{g}{l}}t\right).
\end{equation*}

The steps we took to formulate and solve this problem were:

- Forming a mathematical description of the physical process and writing down the resulting ODE.
- Solving the ODE for a general solution containing arbitrary constants (in this case, a single constant).
- Introducing the initial conditions (in this case two initial
  conditions) and using it to arrive at a particular solution by specifying
  the arbitrary constants.

This course will concentrate on steps 2 and 3. We will concentrate on
techniques for solving equations and not worry too much about existence and
uniqueness of solutions: we will take for granted the usual mathematical
results. 

In this course, only algebraic solution techniques will be covered. We will
neglect the numerical solution of differential equations completely.
