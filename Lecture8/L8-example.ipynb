## Example (Part of an old exam question from a 2nd year Chemistry module)

Green's Theorem relates the line integral of a function $\mathbf{F}=P\mathbf{i}+Q\mathbf{j}$ around a simple closed curve $\gamma$ to an integral over the enclosed area via

\begin{equation*}
\oint_{\gamma}(P\mathrm{d}x+Q\mathrm{d}y)=\iint_S\left(\frac{\partial Q}{\partial x}
 -\frac{\partial P}{\partial y}\right)\mathrm{d}x\mathrm{d}y.
\end{equation*}

For the curve $\gamma$ indicated in the figure below and $P = y/x$ and $Q=x^2 y^2$ evaluate the left and right hand sides of Green's Theorem and show that they are equal.

First, lets evaluate the LHS. Writing the integrals along the lower side of the polygon, right, upper and left side as $I_1$, $I_2$, $I_3$ and $I_4$ respectively, gives

\begin{equation*}
     I_1=\int_{0}^{5}\frac{0}{x}\mathrm{d}x=0,
\end{equation*}

and note $y$ is a constant and hence $\mathrm{d}y$ term will also vanish. Along $I_2$ $x=5$ and $y$ varies from 0 to 3 and hence $P=y/5$, $Q=25y^2$ and hence

\begin{equation*}
     I_2=\int_{5}^{5}\frac{y}{5}\mathrm{d}x+\int_0^3 25y^2\mathrm{d}y=0+225=255,
\end{equation*}

\begin{equation*}
     I_3=\int_5^2 \frac{3}{x}\mathrm{d}x=3\ln\left(\frac{2}{5}\right).
\end{equation*}
    
Along $I_4$, $y=1.5x$ and hence $\mathrm{d}y=1.5\mathrm{d}x$ which gives

\begin{align*}
     I_4&=\int_{2}^{0}\frac{1.5x}{x}\mathrm{d}x+\int_{2}^{0}2.25x^4\times 1.5\mathrm{d}x \\
        &=\frac{3}{2}\int_2^0\left(1+\frac{9}{4}x^4\right)\mathrm{d}x=-\frac{123}{5}.
\end{align*}

Hence,

\begin{equation*}
     LHS = 225+3\ln\left(\frac{2}{5}\right)-\frac{123}{5}.
\end{equation*}
    
For the RHS we have

\begin{equation*}
 \frac{\partial Q}{\partial x}=2xy^2, \qquad \frac{\partial P}{\partial y}=\frac{1}{x}
\end{equation*}

and hence the function we must integrate over the polygon is given by

\begin{equation*}
 2xy^2-\frac{1}{x}.
\end{equation*}

It's simplest to split the integral up into two separate integrals, one over the triangle section, and one over the square. Integrating the triangle section gives
    
\begin{align*}
     \int_{0}^2\left[\int_0^{1.5x}\left(2xy^2-\frac{1}{x}\right)\mathrm{d}y\right]\mathrm{d}x
      &=\int_{0}^{2}\left(2x\left[\frac{y^3}{3}\right]_{0}^{1.5x}-\frac{1}{x}[y]_{0}^{1.5x}\right)\mathrm{d}x \\
      &= \int_{0}^{2}\left[\frac{2\times 1.5^3}{3}x^4-1.5\right]\mathrm{d}x \\
      &=\frac{57}{5}.
\end{align*}

The rectangle gives

\begin{equation*}
     \int_{2}^5\left[\int_0^{3}\left(2xy^2-\frac{1}{x}\right)\mathrm{d}y\right]\mathrm{d}x=225-36+3\ln\left(\frac{2}{5}\right).
\end{equation*}

Hence

\begin{equation*}
     RHS=225+3\ln\left(\frac{2}{5}\right)-\frac{123}{5},
\end{equation*}

as required.
