## Power series re-cap

In the previous lecture we saw that functions can be represented as power series of the form:

\begin{equation*}
  f(x) = a_0 + a_1x + a_2x^2 + \ldots + a_n x^n + \ldots,
\end{equation*}

or formally:

\begin{equation}
  f(x) = \sum_{k=0}^{\infty} a_k x^k,
\end{equation}

where $a_n$ are constants. The above series can also be expressed in a slightly different form:

\begin{equation*}
  f(x)=b_0+b_1(x-x_0)+b_2(x-x_0)^2+\ldots+b_n(x-x_0)^n+\ldots.
\end{equation*}

Notice that both representations are the same, we are only specifying the constants in a different way, e.g.:

\begin{equation*}
  a_0=b_0-b_1x_0+b_2x_0^2-b_2x_0^3+\ldots,
\end{equation*}

and so forth. This second form however allows us to see the following in a more straightforward manner:
- Since $f(x_0)=b_0$ this means that $b_0$ **must be** $f(x_0)$!!
- If we differentiate we get

\begin{equation*}
  f'(x)=b_1+2b_2(x-x_0)+\ldots+nb_n(x-x_0)^{n-1}+\ldots.
\end{equation*}

and hence $f'(x_0)=b_1$!!
- Repeating this process we see that $f''(x_0)=2b_2$, $f'''(x_0)=3\times 2b_3$ and eventually arrive at

\begin{equation*}
  \begin{split}
    f(x)&=f(x_0)+f'(x_0)(x-x_0)+\frac{f''(x_0)}{2!}(x-x_0)^2+\ldots
          +\frac{f^{(n)}(x_0)}{n!}(x-x_0)^n+\ldots\\
        &=\sum_{k=0}^\infty \frac{f^{(k)}(x_0)}{k!}(x-x_0)^k
  \end{split}
\end{equation*}

Thus, we arrive at one form of the **Taylor series** and one of the most important series expansions to be familiar with:

\begin{equation*}
  \begin{split}
    f(x)=\sum_{k=0}^\infty \frac{f^{(k)}(x_0)}{k!}(x-x_0)^k
  \end{split}
\end{equation*}

If let $x\rightarrow x+x_0$ in the above expansion, we get another common form of the Taylor series:

\begin{equation*}
  \begin{split}
    f(x+x_0)=\sum_{k=0}^\infty \frac{f^{(k)}(x_0)}{k!}x^k
  \end{split}
\end{equation*}

(**Note that since $x_0$ is a constant it remains unmodified.**)

Finally, if in either of the above expansions we set $x_0=0$ we end up with the following form of the Taylor series:

\begin{equation*}
  \begin{split}
    f(x)=\sum_{k=0}^\infty \frac{f^{(k)}(0)}{k!}x^k,
  \end{split}
\end{equation*}

which is also known as the Maclaurin series. Note that the expansions will be *most accurate* close to the point $x=x_0$ (hence the Maclaurin series is most accurate near to $x=0$). How exactly we determine this accuracy is beyond the scope of this course (but was considered in the 'supplementary' exercises at the end of last weeks lecture).

## *Examples* (exercise from last week):

### (a) Find the Taylor series about $x=0$ (Maclaurin series) and radii of convergence of the following functions

### (i)

 \begin{equation*}
  f(x)=\frac{1}{1+2x},
 \end{equation*}

### (ii) 
 \begin{equation*}
  g(x)=\frac{1}{2-3x}.
 \end{equation*}

Solutions should state the first four non-zero terms or be presented in terms of a general expression for the $n$-th term.

#### Answer

(i) Evaluating f(0) and its derivatives we have

\begin{align*}
 f(0) &= 1, \\
 f'(0) &= -\left.\frac{2}{(1+2x)^2}\right|_{x=0}=-2 \\
 f''(0) &= \left.\frac{2\times 2\times 2}{(1+2x)^3}\right|_{x=0}=8 \\
 f'''(0) &= -\left.\frac{3\times 2\times 2\times 2 \times 2}{(1+2x)^4}\right|_{x=0}=48 \Rightarrow \\
 \frac{f^{(n)}(0)}{n!} &= (-2)^{n}.
\end{align*}

The Maclaurin series for part (i) is therefore
\begin{equation*}
 f(x)=\sum_{n=0}^{\infty}(-2x)^n,
\end{equation*}
 
The radius of convergence will be given by

\begin{equation*}
 R=\left|\frac{a_n}{a_{n+1}}\right|=\left|\frac{2^n}{2^{n+1}}\right|=\frac{1}{2}.
\end{equation*}

For part (ii),

\begin{align*}
g(0) &= \frac{1}{2}, \\
g'(0) &= -\left.\frac{3}{(2-3x)^2}\right|_{x=0}=\frac{3}{4} \\
g''(0) &= \left.\frac{3\times 3\times 2 \times 2}{(2-3x)^3}\right|_{x=0}=\frac{9}{8}\times 2 \\
 g'''(0) &= \left.\frac{3\times 3\times 3\times 3 \times 2}{(2-3x)^4}\right|_{x=0}=\frac{27}{16}\times 3 \times 2 \Rightarrow \\
 \frac{g^{(n)}(0)}{n!} &= \frac{3^{n}}{2^{n+1}}.
\end{align*}

and therefore:

\begin{equation*}
 g(x)=\frac{1}{2}\sum_{n=0}^{\infty}\left(\frac{3}{2}x\right)^n,
\end{equation*}
 
and the radius of convergence

\begin{equation*}
 R=\left|\frac{a_n}{a_{n+1}}\right|=\left|\frac{3^{n}2^{n+1}}{2^{n}3^{n+1}}\right|=\frac{2}{3}.
\end{equation*}

### (b) Hence, or otherwise, find the Maclaurin series and radius of convergence of the function

\begin{equation*}
 h(x)=\frac{x(3-x)}{2+x-6x^2}.
\end{equation*}

#### Answer

We can show that
\begin{equation*}
 h(x)=\frac{x(3-x)}{2+x-6x^2}=x\left(\frac{1}{1+2x}+\frac{1}{2-3x}\right).
\end{equation*}
 
Hence (using the series from 1.) the series is given by

\begin{equation*}
 h(x)=\sum_{n=0}^{\infty}\left[(-2)^n+\frac{1}{2}\left(\frac{3}{2}\right)^n\right]x^{n+1}.
\end{equation*}
 
and hence the radius of convergence is giveb by $R=1/2$ (the smallest of the two series).

## Ordinary Differential Equations IV: Series solutions to ODEs

We will start with a trivial example:

\begin{equation*}
  y'-y=0
\end{equation*}

We already know that this has the solution:

\begin{equation*}
  y=Ae^x
\end{equation*}

Ignore this for the moment and simply assume that there is a solution which
can be expressed as a power series about zero:

\begin{equation*}
  \begin{split}
    y&=\sum_{n=0}^\infty a_nx^n\\
    &=a_0+a_1x^1+a_2x^2+\ldots+a_nx^n
  \end{split}
\end{equation*}

Differentiating this we have:

\begin{equation*}
  y'=a_1+2a_2x+3a_3x^2+\ldots+na_nx^{n-1}\ldots
\end{equation*}

We can now substitute these series into the original equation:

\begin{gather*}
  y'-y=0\\
  (a_1-a_0)+(2a_2-a_1)x+(3a_3-a_2)x^2+\ldots+((n+1)a_{n+1}-a_{n})x^{n}+\ldots=0
\end{gather*}

For this equality to hold, the coefficient of every term needs to be zero:

\begin{equation*}
\begin{array}{lllll}
  a_1-a_0&=0 &\Rightarrow & a_1&=a_0\\
  2a_2-a_1&=0 &\Rightarrow & a_2&=\frac{a_1}{2}=\frac{a_0}{2}\\
  3a_3-a_2&=0 &\Rightarrow & a_3&=\frac{a_2}{3}=\frac{a_0}{3!}\\
  (n+1)a_{n+1}-a_n&=0&\Rightarrow & a_{n+1}&=\frac{a_n}{(n+1)}=\frac{a_0}{(n+1)!}
\end{array}
\end{equation*}

This allows us to write the power series for $y$ only in terms of $a_0$:

\begin{equation*}
  \begin{split}
    y&=a_0+a_0x+\frac{a_0}{2!}x^2+\frac{a_0}{3!}x^3+\ldots+\frac{a_0}{n!}x^n\ldots\\
    &=a_0(1+x+\frac{x^2}{2!}+\frac{x^3}{3!}+\ldots+\frac{x^n}{n!}\ldots)\\
    &=a_0e^x
  \end{split}
\end{equation*}

## Frobenius' Method

Consider now the general homogeneous linear second order ODE:

\begin{equation*}
  y''+p(x)y'+q(x)y=0
\end{equation*}

Frobenius' Method is a series solution technique for this equation provided
that $(x\times p)(0)$ and $(x^2\times q)(0)$ are both finite.

In the previous example we expanded $y$ in terms of a power series. Here we
will assume a somewhat more general solution:

\begin{equation*}
  \begin{split}
    y&=\sum_{k=0}^\infty a_k x^{m+k}\\ 
    &=x^m\left(\sum_{k=0}^\infty a_kx^k\right)
  \end{split}
\end{equation*}

We do not demand that $m$ is necessarily an integer.

For example, we will attempt to solve:

\begin{equation*}
  4x{y}''+2{y}'+y=0
\end{equation*}

We assume:

\begin{equation*}
  y=\sum_{k=0}^\infty a_k x^{m+k}
\end{equation*}

so that:
\begin{equation*}
  {y}'=\sum_{k=0}^\infty (m+k)a_k x^{m+k-1}
\end{equation*}

and:

\begin{equation*}
  {y}''=\sum_{k=0}^\infty (m+k)(m+k-1)a_k x^{m+k-2}
\end{equation*}

Substituting, we have:

\begin{equation*}
  4\sum_{k=0}^\infty (m+k)(m+k-1)a_k x^{m+k-1}
 + 2\sum_{k=0}^\infty (m+k)a_k x^{m+k-1} + \sum_{k=0}^\infty a_k x^{m+k}=0
\end{equation*}

We need to be able to group the coefficients of each power of $x$. Lets re-write this as:

\begin{equation*}
  \sum_{k=0}^{\infty}\left[4(m+k)(m+k-1)+2(m+k)\right]a_kx^{m+k-1}+\sum_{r=0}^\infty a_r x^{m+r}=0.
\end{equation*}

and hence

\begin{equation*}
 2\sum_{k=0}^\infty (m+k)(2m+2k-1)a_k x^{m+k-1} + \sum_{r=0}^\infty a_r x^{m+r}=0,
\end{equation*}

Notice that we have also changed the index on the second sum. Now, in order
to align the powers of $x$ between the two remaining sums, we extract the
first term from the first sum:

\begin{equation*}
  2m(2m-1)a_0x^{m-1}+  2\sum_{k=1}^\infty (m+k)(2m+2k-1)a_k x^{m+k-1} +
  \sum_{r=0}^\infty a_r x^{m+r}=0
\end{equation*}

Now we can adjust the index on the first sum by substituting $k=r+1$ to
align the powers of $x$:

\begin{gather*}
  2m(2m-1)a_0x^{m-1}+  2\sum_{r=0}^\infty (m+r+1)(2m+2r+1)a_{r+1} x^{m+r} +
  \sum_{r=0}^\infty a_r x^{m+r}=0\\
  2m(2m-1)a_0x^{m-1}+ \sum_{r=0}^\infty \big(2(m+r+1)(2m+2r+1)a_{r+1}+a_r\big)
  x^{m+r} =0 
\end{gather*}

Consider the coefficient occurring at powers of $x^{m-1}$ first (assuming $a_0\ne0$):

\begin{equation*}
  2m(2m-1)=0
\end{equation*}

so $m=1/2$ or $0$. Consider first the case $m=0$. Now for each $r\ge 0$:

\begin{gather*}
  2(m+r+1)(2m+2r+1)a_{r+1}+a_r=0\\
  2(r+1)(2r+1)a_{r+1}+a_r=0\\
  a_{r+1}=-\frac{a_r}{(2r+2)(2r+1)}
\end{gather*}

So:

\begin{align*}
  a_1&=-\frac{a_0}{2}\\
  a_2&=-\frac{a_1}{4\times3}=\frac{a_0}{4!}\\
  a_3&=-\frac{a_2}{6\times5}=-\frac{a_0}{6!}\\
  a_{n}&=(-1)^n\frac{a_0}{(2n)!}
\end{align*}

This gives the power series solution:

\begin{equation*}
  \begin{split}
    y&=x^0\left(a_0-\frac{a_0x}{2!}+\frac{a_0x^2}{4!}-\frac{a_0x^3}{6!}+\ldots\right)\\
    &=a_0\left(1-\frac{(\sqrt{x})^2}{2!}+\frac{(\sqrt{x})^4}{4!}
      -\frac{(\sqrt{x})^6}{6!}+\ldots\right)\\
    &=a_0\cos\sqrt{x}
  \end{split}
\end{equation*}

Now we need to consider the case $m=1/2$. We will perform the same
substitution but to make it clear this is a different series from the last
one, we will write the series coefficients $b_n$.

\begin{gather*}
  2(m+r+1)(2m+2r+1)b_{r+1}+b_r=0\\
  2(r+3/2)(2r+2)b_{r+1}+b_r=0\\
  b_{r+1}=-\frac{b_r}{(2r+3)(2r+2)}
\end{gather*}

So:

\begin{align*}
  b_1&=-\frac{b_0}{3\times2}=-\frac{b_0}{3!}\\
  b_2&=-\frac{b_1}{5\times4}=+\frac{b_0}{5!}\\
  b_3&=-\frac{b_2}{7\times6}=-\frac{b_0}{7!}\\
  b_{n}=(-1)^n\frac{b_0}{(2n+1)!}
\end{align*}

Substituting back into the series, this gives:

\begin{align*}
  y&=x^{1/2}\left(b_0-\frac{b_0x}{3!}+\frac{b_0x^2}{5!}-\frac{b_0x^3}{7!}+\ldots\right)\\ 
  &=b_0\left(\sqrt{x}-\frac{\sqrt{x}^3}{3!}
            +\frac{(\sqrt{x})^5}{5!}-\frac{(\sqrt{x})^5}{7!}+\ldots
       \right)\\
  &=b_0\sin\sqrt{x}
\end{align*}

This produces a general solution to the equation:

\begin{equation*}
  y=a_0\cos\sqrt{x}+b_0\sin\sqrt{x}
\end{equation*}

Let's check that this solution satisfies the equation:

 \begin{gather*}
   y=a_0\cos\sqrt{x}+b_0\sin\sqrt{x}\\
   y'=-\frac{1}{2}\frac{a_0}{\sqrt{x}}\sin\sqrt{x}
   +\frac{1}{2}\frac{b_0}{\sqrt{x}}\cos\sqrt{x}\\
   y''=\frac{1}{4}\frac{a_0}{x^{3/2}}\sin\sqrt{x}
   -\frac{1}{4}\frac{a_0}{\sqrt{x}}\frac{1}{\sqrt{x}}\cos\sqrt{x}
   -\frac{1}{4}\frac{b_0}{x^{3/2}}\cos\sqrt{x}
   -\frac{1}{4}\frac{a_0}{\sqrt{x}}\frac{1}{\sqrt{x}}\sin\sqrt{x}
 \end{gather*}

We can now write out each of the terms in the equation:

\begin{align*}
  4xy''&=\phantom{-}\frac{a_0}{\sqrt{x}}\sin\sqrt{x}-a_0\cos\sqrt{x}-
  \frac{b_0}{\sqrt{x}}\cos\sqrt{x}-a_0\sin\sqrt{x}\\
  2y'&=-\frac{a_0}{\sqrt{x}}\sin\sqrt{x}\phantom{-a_0\cos\sqrt{x}}+
  \frac{b_0}{\sqrt{x}}\cos\sqrt{x}\phantom{-a_0\sin\sqrt{x}}\\
  y&=\phantom{-\frac{a_0}{\sqrt{x}}\sin\sqrt{x}-}a_0\cos\sqrt{x}\phantom{-
  \frac{b_0}{\sqrt{x}}\cos\sqrt{x}}+a_0\sin\sqrt{x}
\end{align*}

It is clear that these terms sum to zero.

#### Properties of Frobenius' Method

The two solutions produced may not necessarily be linearly independent, in
which case the method of variation of parameters can be used to produce a
second solution:

\begin{equation*}
  y=u(x)y_1(x)
\end{equation*}

where $y_1(x)$ is the solution found by Frobenius' method.

In many cases there is no closed analytical expression for the solution. In
this case there are two options open. First, the solution may be left as a
series. Second, it may be possible to express the solution in terms of
certain special functions, e.g. Bessel functions, Legendre functions and Schottky-Klein Prime functions and so forth, which are only expressible as series but which have well understood
properties.

Aside from the methods covered in this course, there are many, many more methods available for solving ODEs, here we have surveyed the core techniques.

## Special functions (Non-examinable)

### Bessel function's of the first kind

Let's look at **Bessel's Ordinary Differential Equation**:

\begin{equation}
  x^2{y}''+x{y}'+(x^2-r^2)y=0,
\end{equation}

for an arbitrary real constant $r$.

Applying Frobenius' method, we end up with $m=\pm r$. For $m=r$, the
solution is:

\begin{equation*}
  y=a_0x^r\left[1-\frac{x^2}{2(r+2)}+\frac{x^4}{2\times4(2r+2)(2r+4)}-\cdots\right].
\end{equation*}

There is no closed form function that corresponds to this series. However,
this series appears a lot in mathematical physics (vibrations, heat
conduction, other problems with cylindrical symmetry) so it would be useful
to be able to work with solutions of this form easily.

The answer to this is to define a new function, called the **Bessel function
of the first kind**:

\begin{equation}
  J_r(x)=\sum_{k=0}^\infty \frac{(-1)^k}{k!\Gamma(r+k+1)}\left(\frac{x}{2}\right)^{r+2k}
\end{equation}

where $\Gamma$ is a generalisation of the factorial function to non-integer numbers,
which we will study shortly.

Using this newly defined function, the general solution to Bessel's equation is:

\begin{equation}
  y=AJ_r(x)+BJ_{-r}(x)
\end{equation}

provided that $r$ is not an integer. (The function $\Gamma$ is not defined
for negative integers).

These functions are not only useful in writing down the solution of the ODE,
but they can be manipulated like any other analytical function. They have
specific properties and there exist particular relationships between, for
example, $J_r$ for different values of the index $r$. There is a whole
branch of mathematics devoted to the study of functions defined in terms of
power series and the results are used in many branches of science and
engineering.

Let's look at the case $r=0$. We will show in the next section
that $\Gamma(k+1) = k!$. The above equation becomes:

\begin{align*}
  J_0(x)& = \sum_{k=0}^{\infty}\frac{(-1)^k}{k!k!}\left(\frac{x}{2}\right)^{2k}\\
        & = 1 - \frac{x^2}{1!1! 2^2} + \frac{x^4}{2!2! 2^4} - \cdots
\end{align*}

This produces a new form of damped oscillation:

```python
%matplotlib inline
import numpy as np
from scipy.special import jv
import matplotlib.pyplot as plt

N = 1000
x = np.linspace(0,20,N)

y0 = jv(0, x)
y1 = jv(1, x)
y2 = jv(2, x)

plt.plot(x, y0, label = "$J_0(x)$")
plt.plot(x, y1, label = "$J_1(x)$")
plt.plot(x, y2, label = "$J_2(x)$")
plt.xlabel("x")
plt.ylabel("y")
plt.ylim((-1.2, 1.2))
plt.legend()
plt.show()
```

### The Gamma function

The vast majority of integrals cannot be evaluated in terms of closed
analytic functions. For example, what is:

\begin{equation*}
  \int e^{-x^2} \mathrm{d}x \textrm{?}
\end{equation*}

Some of these integrals are so common to that new functions have been
defined in terms of their integrals. A notable case in point is the $\Gamma$-function:

\begin{equation*}
  \Gamma(x)=\int_0^\infty t^{x-1}e^{-t}\mathrm{d}t.
\end{equation*}

Having defined the function, we can attempt to study its properties. For
example, we can observe that $\Gamma$ is defined in terms of the integral of
a product so it may be interesting to apply integration by parts. Define:

\begin{equation*}
  u=t^x \qquad \frac{\mathrm{d}v}{\mathrm{d}t}=e^{-t},
\end{equation*}

then we can write:

\begin{equation*}
  \frac{\mathrm{d}u}{\mathrm{d}t}=xt^{x-1} \qquad v=-e^{-t}.
\end{equation*}

Hence:

\begin{align*}
  \Gamma(x+1)&=\left.-t^xe^{-t}\right|_0^\infty +\int_0^\infty
  x t^{x-1}e^{-t}\mathrm{d}t\\
  &=x\int_0^\infty t^{x-1}e^{-t}\mathrm{d}t\\
  &=x\Gamma(x)
\end{align*}

This defines a recurrence relationship. Let's assume $x$ is some positive
integer $n$:

\begin{align*}
  \Gamma(n+1)&=n\Gamma(n)\\
  &=n(n-1)\Gamma(n-1)\\
  &=n!\Gamma(1)
\end{align*}

So what is $\Gamma(1)$?

\begin{align*}
  \int_{0}^{\infty}e^{-t}\mathrm{d}t &=\left[-e^{-t}\right]_0^{\infty}\\
  &=1
\end{align*}

So $\Gamma(n+1)=n!$ justifying our earlier claim that $\Gamma$ is a
generalisation of the factorial function.

Now let's consider $\Gamma(1/2)$:

\begin{align*}
  \Gamma\left(\frac{1}{2}\right)&=\int_0^\infty t^{-\frac{1}{2}}e^{-t}\mathrm{d}t
\end{align*}

We can evaluate this integral using the change of coordinates $u^2=t$. This
results in $2u\mathrm{d}u =\mathrm{d}t$:

\begin{align*}
  \Gamma\left(\frac{1}{2}\right)&=\int_0^\infty \frac{1}{u}e^{-u^2}2u\mathrm{d} u\\
  &=2\int_0^{\infty} e^{-u^2}\mathrm{d} u\\
  &=2\frac{\sqrt{\pi}}{2}\\
  &=\sqrt{\pi}
\end{align*}

The evaluation of this last integral depends on the definition of another
special function, $\mathrm{erf}$, to which we will return soon. Using
the recurrence relation, this result also enables us to obtain
$\Gamma(3/2)$, $\Gamma(5/2)$, $\ldots$

More generally, tables of $\Gamma$ function values exist and it is included
in Python, Matlab, Excel and many other packages.

Lets plot $\Gamma$ for $x>0$:

```python
%matplotlib inline
import numpy as np
from scipy.special import gamma, factorial
import matplotlib.pyplot as plt

x = np.linspace(-3.5, 5.5, 2251)

y = gamma(x)

plt.plot(x, y, 'b', alpha=0.6, label='gamma(x)')

k = np.arange(1, 7)
plt.plot(k, factorial(k-1), 'k*', alpha=0.6, label='(x-1)!, x = 1, 2, ...')
plt.xlim(-3.5, 5.5)
plt.ylim(-10, 25)
plt.grid()
plt.xlabel('x')
plt.legend(loc='lower right')
plt.show()
```

Note the behaviour of $\Gamma$ as x tends to $0$. Using the recurrence relation we
have:

\begin{equation*}
  \Gamma(0)=\frac{\Gamma(1)}{0}
\end{equation*}

Furthermore, since $\Gamma(x>0)>0$:

\begin{equation*}
  \lim_{x\rightarrow +0} \Gamma(x)=\infty
\end{equation*}

What happens when $x<0$? Well one answer is that the integral defining
$\Gamma$ is undefined (the integral diverges) for $x\leq 0$. However, we can still use the
recurrence relation to generate the $\Gamma$ of a negative argument. Using:

\begin{equation*}
  \Gamma(x)=\frac{\Gamma(x+1)}{x}
\end{equation*}

we might for example, evaluate $\Gamma(-3/2)$:

\begin{align*}
  \Gamma(-3/2)&=\frac{\Gamma(-1/2)}{-3/2}\\
  &=\frac{\Gamma(1/2)}{(-3/2)(-1/2)}\\
  &=\frac{4}{3}\sqrt{\pi}
\end{align*}

The recurrence relation breaks down for $\Gamma(x)$ where $x$ is a negative
integer because $\Gamma(0)$ is undefined. In fact, the $\Gamma$ function
approaches positive or negative infinity near negative integers (as shown in our plot above)/

### The Error function

The final special function which we will (briefly) consider is known as the
error function or $\mathrm{erf}$. This rather peculiar name comes from its
relationship with the normal distribution, which in turn describes the error
in many statistical properties commonly measured in science and engineering
contexts. The error function is defined by:

\begin{equation*}
  \mathrm{erf}(x)=\frac{2}{\sqrt{\pi}}\int_0^x e^{-u^2}\mathrm{d}u.
\end{equation*}

As is always the case with special functions, there is no (known) closed form
solution to this integral. Once again we can integrate the function
numerically using python:

```python
%matplotlib inline
import numpy as np
from scipy.special import erf
import matplotlib.pyplot as plt

N = 1000
x = np.linspace(-10,10,N)

y0 = erf(x)
y1 = erf(x)
y2 = erf(x)

plt.plot(x, y0, label = "$\mathrm{erf}(x)$")
plt.plot(x, y1, label = "$\mathrm{erf}(x)$")
plt.plot(x, y2, label = "$\mathrm{erf}(x)$")
plt.xlabel("x")
plt.ylabel("y")
plt.ylim((-1.2, 1.2))
plt.legend()
plt.show()
```

This plot of the error function illustrates it's value at some key
points. For instance, $\mathrm{erf}(0)=0$ and $\mathrm{erf}(\infty)=1$. Note that since $x$
only appears in the definition of $\mathrm{erf}(x)$ as the bound of an integral, it
makes perfect sense to speak of $\mathrm{erf}(\infty)$, or $\mathrm{erf}(-\infty)$ for that matter.
