## First order non-homogeneous ODEs

We begin this lecture by reviewing the solution procedure for ODEs of the form

\begin{equation*}
 \frac{\mathrm{d}y}{\mathrm{d}x}+f(x)y=r(x).
\end{equation*}

We showed that solutions can be written in the form

\begin{equation*}
 y=y_p+y_h
\end{equation*}

where $y_h$ is the solution to the corresponding homogeneous problem

\begin{equation*}
  \frac{\mathrm{d}y}{\mathrm{d}x}+f(x)y=0,
\end{equation*}

and can hence be written as

\begin{equation*}
  y_h=ce^{-\int f(x)\mathrm{d}x}=cv(x)
\end{equation*}

and $y_p$ is the *particular integral* given by

\begin{equation*}
  y_p=u(x)v(x)
\end{equation*}

with

\begin{equation*}
  u=\int \frac{r(x)}{v}\mathrm{d}x.
\end{equation*}

This solution procedure is often termed *the method of variation of parameters*.

#### *Example:*

Solve

\begin{equation*}
 {x^2}y' + xy + 2=0.
\end{equation*}

First lets re-write this in the form

\begin{equation*}
 y' + \frac{y}{x}=-\frac{2}{x^2},
\end{equation*}

and identify

\begin{equation*}
 f(x)=\frac{1}{x}, \; \; \; \; r(x)=-\frac{2}{x^2}.
\end{equation*}

Hence

\begin{equation*}
 v(x)=e^{-\int f(x)\mathrm{d}x}=e^{-\int x^{-1}\mathrm{d}x}=e^{-\ln(x)}=\frac{1}{x},
\end{equation*}

and therefore

\begin{equation*}
 y_h=\frac{c}{x}.
\end{equation*}

Now that $v(x)$ is known we can solve for $u(x)$:

\begin{equation*}
 u=\int \frac{r(x)}{v}\mathrm{d}x=\int \frac{-2x^{-2}}{x^{-1}}\mathrm{d}x=-2\ln(x),
\end{equation*}

and hence

\begin{equation*}
 y_p=uv=-\frac{2\ln(x)}{x}.
\end{equation*}

Hence, bringing these together we have the general solution

\begin{equation*}
 y=-\frac{2\ln(x)}{x}+\frac{c}{x}
\end{equation*}

***Exercise***: Plug the solution back into the original ODE to double check it is indeed a solution.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

N = 1000
x = np.linspace(00000.1,10,N)

def eval_y(x, c):
    return -2*np.log(x)/x+c/x

y1 = eval_y(x, -2)
y2 = eval_y(x, -1)
y3 = eval_y(x, 0)
y4 = eval_y(x, 1)
y5 = eval_y(x, 2)

plt.plot(x, y1, label = "c=-2")
plt.plot(x, y2, label = "c=-1")
plt.plot(x, y3, label = "c=0")
plt.plot(x, y4, label = "c=1")
plt.plot(x, y5, label = "c=2")
plt.xlabel("x")
plt.ylabel("y")
plt.ylim((-4, 6))
plt.legend()
plt.show()
```

## Linear second order ODEs

The most general form of a linear second order equation takes the form

\begin{equation*}
 \frac{\mathrm{d}^2 y}{\mathrm{d}x^2}+f(x)\frac{\mathrm{d}y}{\mathrm{d}x}+g(x)y=r(x).
\end{equation*}

First, lets review what we've learned about homogeneous equations with constant coefficients, that is, equations of the form

\begin{equation*}
 \frac{\mathrm{d}^2 y}{\mathrm{d}x^2}+b\frac{\mathrm{d}y}{\mathrm{d}x}+cy=0,
\end{equation*}

where $b$ and $c$ are real numbers.

In previous lectures we showed that this equation can be factorized as

\begin{equation*}
 \left(\frac{\mathrm{d}}{\mathrm{d}x}-\lambda_1\right)\left(\frac{\mathrm{d}}{\mathrm{d}x}-\lambda_2\right)y=0.
\end{equation*}

where

\begin{equation*}
 \lambda_{1,2}=-\frac{b}{2}\pm\frac{\sqrt{b^2-4c}}{2}.
\end{equation*}

***Exercie***: Ensure that by expanding the above operator with $\lambda_{1,2}$ as stated we retrieve the original ODE.

We are then left with two first order ODEs of the form

\begin{equation*}
 \frac{\mathrm{d}y}{\mathrm{d}x}-\lambda y = 0,
\end{equation*}

to solve, which yield solutions of the form

\begin{equation*}
 y\propto e^{\lambda x}.
\end{equation*}

Hence, owing to the linearity of the underlying equation we expect general solutions of the form

\begin{equation*}
 y=Ae^{\lambda_1 x} + Be^{\lambda_2 x}.
\end{equation*}

Note that the values of $\lambda_1$ and $\lambda_2$ can be found by substituting the expected form of our solution ($e^{\lambda x}$) back into the differential equation, which gives

\begin{equation*}
 \lambda^2 e^{\lambda x} + b \lambda e^{\lambda x} +c e^{\lambda x} = 0.
\end{equation*}

Dividing through by $e^{\lambda x}$ we are left with a quadratic equation for $\lambda$:

\begin{equation*}
 \lambda^2+b\lambda+c=0,
\end{equation*}

which has solutions

\begin{equation*}
 \lambda_{1,2}=-\frac{b}{2}\pm\frac{\sqrt{b^2-4c}}{2}.
\end{equation*}

just as we showed above!

### Regimes and examples

We'll now look at what solutions PDEs of this type can yield and relate each to a *mechanical system*.

### Two distinct real roots ($b^2>4c$)

That is, the solution takes the form

\begin{equation*}
 y=Ae^{\lambda_1 x}+Be^{\lambda_2 x},
\end{equation*}

with $\lambda_1$, $\lambda_2$ both real.

**Example:**

\begin{equation*}
 \frac{\mathrm{d}^2 y}{\mathrm{d}x^2}+\frac{\mathrm{d} y}{\mathrm{d}x}-6y=0.
\end{equation*}

Using one of the methods above this yield $\lambda_1=3$, $\lambda_2=-2$ (or $\lambda_1=-2$, $\lambda_2=3$ - it doesn't matter how we label these, we'll get equivalent results). Hence

\begin{equation*}
 y=Ae^{3x}+Be^{-2x}
\end{equation*}

How do functions of this kind look? Lets pick some initial conditions (so that we can evaluate $A$ and $B$) and see. Lets take:

\begin{equation*}
 y(0)=0, \; \; \; \; \; \; \frac{\mathrm{d}y}{\mathrm{d}x}(0)=4.
\end{equation*}

Hence,

\begin{equation*}
 A+B=0, \; \; \; \; \; \; 3A-2B=4,
\end{equation*}

which gives

\begin{equation*}
 A=\frac{4}{5}, \; \; \; \; \; \; B=-\frac{4}{5}.
\end{equation*}

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

N = 1000
x = np.linspace(-10,10,N)

def y_2real(x, A, B):
    return A*np.exp(A*x)+A*np.exp(B*x)

y = y_2real(x, 0.8, -0.8)

plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.ylim((0, 10))
plt.legend()
plt.show()
```

As we saw previously, 2nd Order ODEs can also be used to model mechanical systems, one such system being an oscillating mass on a spring:

![SegmentLocal](figures/ms.gif "segment")
**Image:** An undamped spring–mass system undergoes simple harmonic motion. See [here](https://en.wikipedia.org/wiki/Simple_harmonic_motion) for image source.

In this case the differential equation can be written as

\begin{equation*}
 m\frac{\mathrm{d}^2 y}{\mathrm{d}t^2}+c\frac{\mathrm{d} y}{\mathrm{d}t}+ky=0,
\end{equation*}

where $m$ is the mass (or the 'weight'), $c$ is the damping coefficient and $k$ is the *spring constant* which is related to the stiffness of the spring.

**Important**: Notice that since this is a real mechanical system $m$, $c$ and $k$ must all take real values $>0$.

Lets re-write the above as

\begin{equation*}
 \frac{\mathrm{d}^2 y}{\mathrm{d}t^2}+2\alpha\frac{\mathrm{d} y}{\mathrm{d}t}+\omega^2 y=0,
\end{equation*}

where

\begin{equation*}
 \alpha=\frac{c}{2m}, \; \; \; \; \; \; \omega=\sqrt{\frac{k}{m}}.
\end{equation*}

Thus, when we have two distinct real roots, this corresponds to $\alpha^2>\omega^2$ and

\begin{equation*}
 \lambda_{1,2}=-\alpha\pm\sqrt{\alpha^2-\omega^2}.
\end{equation*}

**Important:** Since $\alpha>0$ and real, this means that both values of $\lambda$ **MUST** be negative. This restriction ensures we get sensible results and that our solution doesn't blow up.

**Example**:

Lets consider an example with $m=1$, $c=3$ and $k=2$ with $y(0)=1$ and $\dot{y}(0)=u_0$ which has a solution:

\begin{equation*}
 y = u_0 \left(e^{-t}-e^{-2t}\right)+2e^{-t}+e^{-2t}. 
\end{equation*}

Such oscillators are known as **over-damped**. Lets plot this for a few values of $u_0$:

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

N = 1000
t = np.linspace(0,10,N)

def y_m2r(t, u0):
    return u0*(np.exp(-t)-np.exp(-2*t))+2*np.exp(-t)-np.exp(-2*t)

y1 = y_m2r(t, -1)
y2 = y_m2r(t, 0)
y3 = y_m2r(t, 1)

plt.plot(t, y1, label = "u0=-1")
plt.plot(t, y2, label = "u0=0")
plt.plot(t, y3, label = "u0=1")
plt.xlabel("t")
plt.ylabel("y")
plt.ylim((0, 1.2))
plt.legend()
plt.show()
```

### Two imaginary roots - complex conjugate pair ($4c>b^2$)

From our above analysis, it is clear that complex roots will occur when $b^2-4c<0$ or $4c>b^2$.

The general solution in this case is

\begin{equation*}
 y=Ae^{(p+iq)x}+Be^{(p-iq)x}
\end{equation*}

where
\begin{equation*}
 p=-\frac{b}{2}, \; \; \; \; \; \; q=\frac{\sqrt{4c-b^2}}{2},
\end{equation*}

or

\begin{equation*}
 y=e^{px}\left(Ae^{iqx}+Be^{-iqx}\right)=e^{px}(\hat{A}\cos(qx)+\hat{B}\sin(qx)).
\end{equation*}

**Example**:
    
\begin{equation*}
 \frac{\mathrm{d}^2 y}{\mathrm{d}x^2}+4\frac{\mathrm{d} y}{\mathrm{d}x}+9y=0.
\end{equation*}

Now, $b=4$, $c=9$ and hence:

\begin{equation*}
 \lambda_{1,2}=-2\pm\frac{i\sqrt{4\times 9 - 4^2}}{2}=-2\pm i\sqrt{5}.
\end{equation*}

Solutions therefore take the form:

\begin{equation*}
 y=e^{-2x}(\hat{A}\cos(\sqrt{5}x)+\hat{B}\sin(\sqrt{5}x))
\end{equation*}

When considering mechanical systems with solutions of this form, there are two sub-cases
1. $p=0$
2. $p<0$ (Note we can rule out $p>0$ since this would correspond to the system 'blowing up')

#### case 1: p=0, free-oscillation

This results from oscillatory systems with no damping i.e. $c=0$ and hence $\alpha=0$


\begin{equation*}
 \frac{\mathrm{d}^2 y}{\mathrm{d}t^2}+\omega^2 y=0.
\end{equation*}

**Example**:
    
\begin{equation*}
 \frac{\mathrm{d}^2 y}{\mathrm{d}t^2}+4\pi^2 y=0, \; \; \; \; y(0)=0, \; \; \; \; \frac{\mathrm{d}y}{\mathrm{d}t}(0)=1.
\end{equation*}

Now, $p=0$ and $q=2\pi$ and thus our general solution is

\begin{equation*}
 y(t)=A\cos(2\pi t)+B\sin(2\pi t)
\end{equation*}

Applying our boundary conditions we get:

\begin{equation*}
 A\cos(0)+B\sin(0)=0 \rightarrow A=0,
\end{equation*}

and

\begin{equation*}
 -2\pi A\sin(0)+2\pi B\cos(0)=1 \rightarrow B=\frac{1}{2\pi},
\end{equation*}

and therefore

\begin{equation*}
 y(t)=\frac{1}{2\pi}\sin(2\pi t).
\end{equation*}

Note that this oscillator has an angular velocity $\omega$ of $2\pi$ - that is every 1 unit of time it moves through one full cycle i.e. it's period T=1. A plot:

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

N = 1000
t = np.linspace(0,5,N)

def y_fo(t):
    return 0.5/np.pi*np.sin(2*np.pi*t)

y = y_fo(t)

plt.plot(t, y)
plt.xlabel("t")
plt.ylabel("y")
plt.ylim((-0.2, 0.2))
plt.legend()
plt.show()
```

#### Case 2: Damped oscillator

If we slightly modify the previous example by introducing an $\alpha>0$ but maintaining the relationship $\omega^2 > \alpha^2$ we'll see we now get damped oscillating systems. Lets dive straight into and example:

\begin{equation*}
 \frac{\mathrm{d}^2 y}{\mathrm{d}t^2}+2\sqrt{3}\pi\frac{\mathrm{d}y}{\mathrm{d}t}+4\pi^2 y=0, \; \; \; \; y(0)=0, \; \; \; \; \frac{\mathrm{d}y}{\mathrm{d}t}(0)=u_0.
\end{equation*}

Since $p=-\alpha$ and $q=\sqrt{\omega^2-\alpha^2}$ we obtain

\begin{equation*}
 y(t)=e^{-\sqrt{3}\pi t}(A\cos(\pi t)+B\sin(\pi t)).
\end{equation*}

and applying the boundary conditions:

\begin{equation*}
 y(t)=u_0\frac{e^{-\sqrt{3}\pi t}}{\pi}\sin(\pi t).
\end{equation*}

Notice that the affect of the damping is to reduce the frequency of oscillation (the period is now $T=2$) and to reduce the maximum amplitude over time. A plot of the damped system is:

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

N = 1000
t = np.linspace(0,5,N)

def y_fo(t, u0):
    return u0*np.exp(-np.sqrt(3)*np.pi*t)/np.pi*np.sin(np.pi*t)

y = y_fo(t, 10)

plt.plot(t, y)
plt.xlabel("t")
plt.ylabel("y")
plt.ylim((-0.1, 0.7))
plt.legend()
plt.show()
```

### Repeated roots ($b^2=4c$)

In this case, we have $p=-b/2$ and $q=0$.

Last week, we saw this leads to solutions of the form

\begin{equation*}
 y=(A+Bx)e^{px},
\end{equation*}

Lets take a closer look at an example. Consider the equation:

\begin{equation*}
 \frac{\mathrm{d}^2 y}{\mathrm{d}t^2}+2\frac{\mathrm{d}y}{\mathrm{d}t}+y=0.
\end{equation*}

We can see this equation yields $\lambda_{1,2}=-1$. Hence

\begin{equation*}
 y=(A+Bt)e^{-t}.
\end{equation*}

Lets double check this is indeed a solution (by back substitution)

\begin{equation*}
 \left((A+Bt)e^{-t}-2Be^{-t}\right)+2\left(Be^{-t}-(A+Bt)e^{-t}\right)+\left((A+Bt)e^{-t}\right)=0,
\end{equation*}

It's interesting to ask, why is this case so different from the others? And how can we understand this difference in more detail?

### How, and why, can we be sure we've found all solutions?

This subject can get a little deep, and we don't have time to go over everything in detail. But lets quickly get a flavor of how we can go about proving this. Note that more details regarding this issue can be found in the reading list for this course.

Suppose $y$ is a solution of

\begin{equation*}
 ay''+by'+c=0,
\end{equation*}

and $\lambda_1$ is *one* root of the characteristic equation

\begin{equation*}
 a\lambda^2+b\lambda+c=0.
\end{equation*}

Define a new differentiable function

\begin{equation*}
 u = ay' +(a\lambda_1+b)y,
\end{equation*}

and therefore

\begin{equation*}
 u' = ay'' +(a\lambda_1+b)y',
\end{equation*}

Using the new function $u$ along with our characteristic equation we can show that $u$ satisfies

\begin{equation*}
 u'-\lambda_1u=0.
\end{equation*}

This is a first order ODE which we know has solutions

\begin{equation*}
 u=C_1e^{\lambda_1 x}.
\end{equation*}

Putting this back into our definition of $u$ yields

\begin{equation*}
 ay'+(a\lambda_1+b)y=C_1e^{\lambda_1x}.
\end{equation*}

This is another 1st order linear ODE but now in $y$ and now non-homogeneous. But we know how to solve this, and we'll get

\begin{equation*}
 y=C_2 e^{\lambda_2x}\int e^{(\lambda_1-\lambda_2)x}\mathrm{d}x,
\end{equation*}

where $\lambda_2$ is the *other* root of our characteristic equation. Ah ha, we now have $y$ in a form where on the right hand side we have a $\lambda_1-\lambda_2$ term. We have two distinct cases:

**$\lambda_1 \ne \lambda_2$:**

If we go ahead and solve this we'll end up with solutions of the form

\begin{equation*}
 y=Ae^{\lambda_1 x} + Be^{\lambda_2 x},
\end{equation*}

that is, the solutions we've seen before in this case.

**$\lambda_1 = \lambda_2$:**

In this case our integral reduces to

\begin{equation*}
 y=C_2 e^{\lambda_2x}\int \mathrm{d}x,
\end{equation*}

and hence

\begin{equation*}
 y=C_2 e^{\lambda_2x}(x+C_3)=(A+Bx)e^{\lambda_2x}.
\end{equation*}

Note that to be completely rigorous we'd have to check a few more things, in particular in the case of complex roots, but the above analysis should give you an idea of why the solutions we see that work come about.

<div class="alert alert-block alert-warning"><b>Note:</b> We're now back to examinable material.</div>

Lets now return to our previous example

\begin{equation*}
 \ddot{y}+2\dot{y}+y=0,
\end{equation*}

and add the boundary conditions

\begin{equation*}
 y(0)=1, \; \; \; \; \; \; \; \dot{y}(0)=u_0.
\end{equation*}

We now know that the general solution to this equation is

\begin{equation*}
 y(t) = (A+Bt)e^{-t}.
\end{equation*}

Applying our boundary conditions we get

\begin{equation*}
 y(t)=(1+(1+u_0)t)e^{-t}.
\end{equation*}

Note that this is the 'critically damped' case and the behavior of such mechanical systems is similar to the over-damped case. Lets plot some results for various $u_0$ below:

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

N = 1000
t = np.linspace(0,10,N)

def y_c(t, u0):
    return (1+(1+u0)*t)*np.exp(-t)

y1 = y_c(t, -1)
y2 = y_c(t, 0)
y3 = y_c(t, 1)

plt.plot(t, y1, label = "u0=-1")
plt.plot(t, y2, label = "u0=0")
plt.plot(t, y3, label = "u0=1")
plt.xlabel("t")
plt.ylabel("y")
plt.ylim((0, 1.4))
plt.legend()
plt.show()
```

## Non-homogeneous second order linear ODEs

Having examined the homogeneous second order ODE, we now proceed to the
general second order linear ODE with constant coefficients:

\begin{equation}
  y'' + by' + cy  = r(x)
\end{equation}

Similar to first oder non-homogeneous ODEs the solution can again be written as

\begin{equation}
  y = y_h + y_p
\end{equation}

where $y_h$ is the solution to the homogeneous part of the problem and $y_p$ is a particular solution.

We know how to determine $y_h$, so let us examine some methods of determining $y_p$.

### The method of undetermined coefficients

This technique is relatively simple but may not work in all
circumstances. The idea is to use what we know about the derivatives of
functions to make an assumption about the form that a solution is likely to
take. For example, consider the equation:

\begin{equation}
  y''-4y=3\sin 2x
\end{equation}

Consider the right hand side. The derivatives of $\sin nx$ have the form
$A\sin nx$ and $B\cos nx$ so we look for a particular integral:

\begin{equation}
  y_p=A\sin 2x + B\cos 2x
\end{equation}

If we substitute equation our trial function back into the original
equation and evaluate values for the constants
$A$ and $B$, then the method has given us a particular solution. The approach
of the method of undetermined coefficients is to devise a solution based on the form of $r(x)$, termed **a trial solution**, and see if it works and if we can determine the coefficients.

Taking the second derivative of our trial function, we have:

\begin{equation}
  y_p'' = -4A\sin2x-4B\cos2x
\end{equation}

Back substitution into the original problem yields:

\begin{equation*}
  -8A\sin2x-8B\cos2x=3\sin2x
\end{equation*}

By equating corresponding powers/function of $x$, this produces the solution:

\begin{equation*}
  A=-\frac{3}{8}\quad\quad B=0
\end{equation*}

so the particular solution is:

\begin{equation*}
  y_p=-\frac{3}{8}\sin 2x
\end{equation*}

Note that once again there are no arbitrary constants in the particular
integral as these are subsumed in the solution to the homogeneous part of
the equations.

To complete the solution, we must find the general solution to the
homogeneous problem. The characteristic equation is:

\begin{equation*}
  \lambda^2-4=0
\end{equation*}

which has roots $\lambda=\pm2$ so the solution to the homogeneous part is:

\begin{equation*}
  y_h = c_1 e^{-2x} + c_2 e^{2x}
\end{equation*}

This yields a full solution:

\begin{equation*}
  y = c_1 e^{-2x} + c_2 e^{2x}-\frac{3}{8}\sin 2x
\end{equation*}

Do not confuse the constants of integration $c_1$ and $c_2$ with the (previously)
undetermined coefficients $A$ and $B$. The coefficients $A$ and $B$ are part of
the heuristic approach of finding the particular integral, while the constants of
integration $c_1$ and $c_2$ are part of the homogeneous solution and can only be
determined with initial conditions **after the full solution $y = y_h + y_p$
has been determined.**

We can check the correctness of this solution by substitution:

\begin{gather*}
  y''=4c_1e^{-2x} +4c_2e^{2x}+4\times \frac{3}{8}\sin 2x\\
  -4y=-4c_1e^{-2x}-4c_2e^{2x}+4\times\frac{3}{8}\sin 2x
\end{gather*}

Hence:

\begin{gather*}
  \begin{split}
    y''-4y&=8\times\frac{3}{8}\sin 2x\\
    &=3\sin 2x
  \end{split}
\end{gather*}

#### How to determine the trial function

The table below outlines the form of the trial solution, for various
classes of functions on the right hand side, $r(x)$ in equation:

| ------------------------- $r(x)$ -------------------------| ------------------------- Trial $y_p$ ------------------------- |
| --------------------------- | ------------------------------|
| $k e^{\gamma x}$     | $C e^{\gamma x}$       |
| $k x^n$,  $n = 0,1,\ldots$   | $ K_n x^n + K_{n-1} x^{n-1} + \ldots + K_1 x + K_0$        |
| $k\cos(\omega x)$ or $k\sin(\omega x)$  | $K\cos(\omega x) +M\sin(\omega x)$        |
| $ke^{\alpha x}\cos(\omega x)$ or $ke^{\alpha x}\sin(\omega x)$   | $e^{\alpha x}(K\cos(\omega x)+M\sin (\omega x))$         |

The following guidelines should be used to calculate constants appearing in 
the trial solutions:
- In the simplest of cases the right hand side $r(x)$ is one listed
in the left column of the table above. Substitution
of the trial solution will give the constants of integration.
- If the trial $y_p$ is also a solution to the homogeneous part, the
substitution will yield $0 = r(x)$. Multiply the trial solution by $x$.
- If the trial $y_p$ is also a double root of the characteristic
equation, substitution will yield $0 = r(x)$. Multiply the trial
solution by $x^2$.
- If $r(x)$ is a sum (linear combination) of function in the left
column in the table above, use a linear combination
of the corresponding function in the right column.

## Variation of parameters

The method of variation of parameters is a direct generalisation of the
method for first order ODEs. The slight complication is that we now have two
independent solutions from which to work. If the general solution to the
homogeneous problem is:

\begin{equation*}
  y=Ay_1+By_2
\end{equation*}

then we look for a solution of the form:

\begin{equation*}
  y_p=u(x)y_1+v(x)y_2
\end{equation*}

If we substitute this solution into the differential equation, we will have
one equation for two unknown functions. This enables us to impose another
auxiliary equation, which we choose so as to simplify the derivatives:

\begin{equation*}
  u'y_1+v'y_2=0.
\end{equation*}

Differentiating $y_p$, we have:

\begin{equation*}
  \begin{split}
    y'_p&=u'y_1+uy_1'+v'y_2+vy'_2\\
    &=uy'_1+vy'_2
  \end{split}
\end{equation*}

and thence:

\begin{equation*}
  \begin{split}
    y''_p = u'y'_1 + uy_1'' + v'y_2' + vy_2''
  \end{split}
\end{equation*}

Substituting these into the equation yields:

\begin{equation*}
  u(y_1''+by_1'+cy_1)+v(y_2''+by_2'+cy_2)+u'y_1'+v'y_2'=r
\end{equation*}

Since $y_1$ and $y_2$ are solutions to the homogeneous part of the equation,
the expressions in brackets vanish. This leaves:

\begin{equation*}
  u'y_1'+v'y_2'=r
\end{equation*}

By considering this equation and the auxiliary equation together, we can
construct a two-dimensional linear system to solve for $u'$ and $v'$:

\begin{equation*}
  \begin{bmatrix}
    y_1 & y_2 \\
    y_1' & y_2'
  \end{bmatrix}
  \begin{bmatrix}
    u'\\
    v'\\
  \end{bmatrix}
  =
  \begin{bmatrix}
    0\\
    r
  \end{bmatrix}
\end{equation*}

If we write $\textbf{A}$ for the matrix:
\begin{equation*}
  \textbf{A}=  
  \begin{bmatrix}
    y_1 & y_2 \\
    y_1' & y_2'
  \end{bmatrix}
\end{equation*}

then the inverse of $\textbf{A}$ is given by:

\begin{equation*}
  \textbf{A}^{-1}=
  \frac{1}{\det\textbf{A}}
  \begin{bmatrix}
    y_2' & -y_2 \\
    -y_1' & y_1
  \end{bmatrix}
\end{equation*}

where $\det\textbf{A}$ is the **determinant** of $\textbf{A}$, which is given by
$y_1y_2'-y_2y_1'$. Using this inverse matrix we can write:

\begin{equation*}
  \begin{split}
    \begin{bmatrix}
      u'\\
      v'\\
    \end{bmatrix}
    &=
    \textbf{A}^{-1}
    \begin{bmatrix}
      0\\
      r
    \end{bmatrix}\\
    &=
    \frac{1}{\det\textbf{A}}
    \begin{bmatrix}
      y_2' & -y_2 \\
      -y_1' & y_1
    \end{bmatrix}
    \begin{bmatrix}
      0\\
      r
    \end{bmatrix}\\
    &= \frac{1}{\det\textbf{A}}
    \begin{bmatrix}
      -y_2r\\
      y_1r
    \end{bmatrix}
  \end{split}
\end{equation*}

That is to say:

\begin{equation*}
  u'=\frac{-y_2r}{W} \quad\quad v'=\frac{y_1r}{W}
\end{equation*}

The quantity $W\equiv y_1y_2'-y_2y_1'$, the determinant of $\textbf{A}$ is
known as the **Wronskian** after the Polish philosopher and
mathematician Józef Hoene-Wroński .

Finally we can write expressions for $u$ and $v$:

\begin{equation*}
  u=-\int\frac{y_2r}{W}\mathrm{d}x \quad\quad v=\int\frac{y_1r}{W}\mathrm{d}x
\end{equation*}

In summary, the solution procedure is:
1. Find the homogeneous solutions $y_1$, $y_2$.
2. Calculate $W$, $u$ and $v$.
3. $y_p=uy_1+vy_2$.
4. Form the general solution as $y=y_h+y_p$.

#### Example

Find the general solution to:

\begin{equation*}
  y''-y=e^x
\end{equation*}

**Solve the homogeneous equation:**

The characteristic equation is:

\begin{equation*}
  \lambda^2-1=0
\end{equation*}

The roots of this equation are $\lambda=\pm1$ so the independent solutions
of the homogeneous equation are:

\begin{equation*}
  y_1=e^x\quad\quad y_2=e^{-x}
\end{equation*}

**Calculate $W$, $u$ and $v$:**

\begin{equation*}
  \begin{split}
    W&=e^x\frac{\mathrm{d}}{\mathrm{d}x}{(e^{-x})}-e^{-x}\frac{\mathrm{d}}{\mathrm{d}x}{(e^{x})}\\
    &=-e^xe^{-x}-e^{-x}e^x\\
    &=-2
  \end{split}
\end{equation*}

\begin{equation*}
  \begin{split}  
    u&=-\int\frac{e^{-x}e^x}{-2}\mathrm{d}x\\
    &=\frac{x}{2}
  \end{split}
\end{equation*}

\begin{equation*}
  \begin{split}
    v&=\int\frac{e^{x}e^{x}}{-2}\mathrm{d}x\\
    &=-\frac{e^{2x}}{4}
  \end{split}
\end{equation*}

Substituting back, we have:

\begin{equation*}
  \begin{split}
    y_p&=uy_1+vy_2\\
    &=\frac{x}{2}e^x-\frac{e^x}{4}
  \end{split}
\end{equation*}

Once again, there are no arbitrary constants in the particular
integral. These enter the general solution via the homogeneous part:

\begin{gather*}
  y=Ae^x+Be^{-x}+\frac{x}{2}e^x-\frac{e^x}{4}\\
  y=\hat{A}e^x+Be^{-x}+\frac{x}{2}e^x
\end{gather*}

## Linear ODEs with variable coefficients

The next stage of complexity in the equations is achieved by allowing the
coefficients to be general functions of $x$:

\begin{equation*}
  \frac{\mathrm{d}^2 y}{\mathrm{d}x^2}+b(x)\frac{\mathrm{d}y}{\mathrm{d}x}+c(x)y=r(x)
\end{equation*}

These equations are **much** harder to solve than the constant
coefficient case. A core technique for solution are series approximation
techniques, which we shall examine later. However, some of these equations
can be solved by purely algebraic manipulation.

### The Cauchy or Euler equation

An example of a second order ODE with variable coefficients which is solvable
is the equation variously known as the Cauchy or Euler equation:

\begin{equation*}
  x^2y''+bxy'+cy=0
\end{equation*}

To solve this equation, we first make the substitution $x=e^t$ from which we
have:

\begin{equation*}
  \begin{split}
    \frac{\mathrm{d}y}{\mathrm{d}x}&=\frac{\mathrm{d}y}{\mathrm{d}t}\frac{\mathrm{d}t}{\mathrm{d}x}\\
    &=\frac{\mathrm{d}y}{\mathrm{d}t}e^{-t}\\
    &= \frac{\mathrm{d}y}{\mathrm{d}t}\frac{1}{x}
  \end{split}
\end{equation*}

and:

\begin{equation*}
  \begin{split}
    \frac{\mathrm{d}^2y}{\mathrm{d}x^2}&=\frac{\mathrm{d}}{\mathrm{d}x}\left(\frac{\mathrm{d}y}{\mathrm{d}t}e^{-t}\right)\\
    &=\frac{1}{x}\frac{\mathrm{d}}{\mathrm{d}t}\left(\frac{\mathrm{d}y}{\mathrm{d}t}e^{-t}\right)\\
    &=\frac{1}{x}\left(\frac{\mathrm{d}^2y}{\mathrm{d}t^2}e^{-t}-\frac{\mathrm{d}y}{\mathrm{d}t}e^{-t}\right)\\
    &=\frac{1}{x^2}\left(\frac{\mathrm{d}^2y}{\mathrm{d}t^2}-\frac{\mathrm{d}y}{\mathrm{d}t}\right)
  \end{split}
\end{equation*}

Substituting the derivatives back into the equation, we have:

\begin{equation*}
  \frac{\mathrm{d}^2y}{\mathrm{d}t^2}+(b-1)\frac{\mathrm{d}y}{\mathrm{d}t}+cy=0
\end{equation*}

which has constant coefficients and characteristic equation:

\begin{equation*}
  \lambda^2+(b-1)\lambda+c=0
\end{equation*}

We will examine just a couple of cases. Suppose the roots of the
characteristic equation are real. Then we have the general solution:

\begin{equation*}
  y=Ae^{\lambda_1t}+Be^{\lambda_2t}
\end{equation*}

or

\begin{equation*}
  y=Ax^{\lambda_1}+Bx^{\lambda_2}
\end{equation*}

Alternatively, suppose we have a double root. Then:

\begin{equation*}
  y=(A+Bt)e^{\frac{1-b}{2}t}
\end{equation*}

but, $x=e^t$ gives $x^{\frac{1-b}{2}} = e^{\frac{1-b}{2}t}$ and $t = \ln x$, so:

\begin{equation*}
  y=(A+B \ln x)x^{\frac{1-b}{2}}
\end{equation*}

#### Example

Solve the Cauchy equation:

\begin{equation*}
  x^2y'' - 2.5xy'-2y=0
\end{equation*}

This yields a characteristic equation:

\begin{equation*}
  \lambda^2-3.5\lambda-2y=0
\end{equation*}

Note that the coefficient of $\lambda$ is $-2.5-1=-3.5$. This equation has
real roots $\lambda_1=4$, $\lambda_2=-0.5$ so that the general solution is:

\begin{equation*}
  y=Ax^4+\frac{B}{\sqrt{x}}
\end{equation*}
