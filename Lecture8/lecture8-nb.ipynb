# Lecture 8 - Multiple Integrals

## Calculus of functions of two variables

### Revision of calculus of functions of a single variable

We are already very familiar with the calculus of functions of a single
variable. If we write $f(x)$ for some function then the gradient of this
function is given by:

\begin{equation*}
  \frac{\mathrm{d}f}{\mathrm{d}x}\equiv f'(x)
\end{equation*}

The gradient is the slope of the tangent line to the curve. In the
following figure, the tangent line shown has gradient $f'(x)|_{x=a}$, or $f'(a)$.

<img src='./figures/d1.png' width=400>

Similarly, the integral of a function is the area under its graph. We
write:

\begin{equation*}
  \int_a^b f(x)\mathrm{d} x
\end{equation*}

Which is illustrated by the following plot:

<img src='./figures/d2.png' width=400>

For example, consider a body starting at rest and falling under gravity.
Then the velocity of that body is given by:

\begin{equation*}
  u(t)=-gt
\end{equation*}

The acceleration is the rate of change of velocity which is given by:

\begin{equation*}
  \frac{\mathrm{d}u}{\mathrm{d}t}=-g
\end{equation*}

While the displacement of the body during the time interval $(t_1,t_2)$ is:

\begin{equation*}
  \int_{t_1}^{t_2}u\mathrm{d}x=\left.\frac{gt^2}{2}\right|_{t_1}^{t_2}
\end{equation*}

## Two variables

In two dimensions, a function defines a surface which has a tangent plane at
each point. The slope of the tangent plane is different in different
directions, which we notate using partial derivatives:

\begin{equation*}
   \left. \frac{\partial f}{\partial y} \right|_x \quad \textrm{or} \left. \quad \frac{\partial f}{\partial x} \right|_y
\end{equation*}

For example, consider the function:

\begin{equation*}
  P=\frac{nRT}{V}
\end{equation*}

Then:

\begin{align*}
  \left. \frac{\partial P}{\partial T} \right|_V &=\frac{nR}{V}\\
  &=\frac{P}{T}
\end{align*}

but:

\begin{align*}
  \left. \frac{\partial P}{\partial V} \right|_T &=-\frac{nR}{V^2}\\
  &=-\frac{P}{V}
\end{align*}

Not only are these two functions different, they can be functions of the
variables held constant in the derivatives.

How would one integrate functions of two variables? There are a number of
ways but the way we are interested in is to integrate with respect to one
variable and then with respect to the other:

\begin{equation*}
  \int\left[\int f(x,y)\mathrm{d} y\right]\mathrm{d} x
\end{equation*}

In this case, we integrate with respect to $y$ first, then with respect to
$x$. As a form of shorthand notation, we write:

\begin{equation*}
  \int\int f(x,y)\mathrm{d}y\mathrm{d}x
\end{equation*}

This is known as a **double integral**, but what does this mean?

As an example, let's consider:

\begin{equation*}
  \int_0^1\int_0^1 f(x,y)\,\mathrm{d}y\,\mathrm{d}x
\end{equation*}

Consider first the inner integral:

\begin{equation*}
  \int_0^1 f(x,y)\mathrm{d}y
\end{equation*}

If we hold the value of $x$ constant then we can perform this integral along
that line of constant $x$:

\begin{equation*}
  \int_0^1 f(x,y)\mathrm{d}y=F(x)
\end{equation*}

Hence, the integral represents the area of the vertical section at $x$=constant extending from the $x-y$ plane to the surface $f(x,y)$, between $y=0$ and $y=1$. This integral, $F(x)$, is a function of the value of $x$. I can now integrate the volume under the surface $f(x,y)$ by integrating $F(x)$:

\begin{equation*}
  \int_0^1 F(x)\mathrm{d}x
\end{equation*}

Hence, the double integral represents the volume under the surface $f(x,y)$ over the region of integration in the $x-y$ plane. 

### Example

Evaluate:

\begin{align*}
  \int_0^1\int_0^1x^2y\,\mathrm{d}y\,\mathrm{d}x 
  &= \int_0^1\left.\frac{x^2y^2}{2}\right|_0^1\mathrm{d}x\\
  &=\int_0^1\frac{x^2}{2}\mathrm{d}x\\
  &=\left.\frac{x^3}{6}\right|_0^1\\
  &=\frac{1}{6}
\end{align*}

So what happens if we reverse the order of integration:

\begin{align*}
  \int_0^1\int_0^1x^2y\,\mathrm{d}x\,\mathrm{d}y 
  &=  \int_0^1\left.\frac{x^3y}{3}\right|_0^1\mathrm{d}y\\
  &=\int_0^1\frac{y}{3}\mathrm{d}y\\
  &=\left.\frac{y^2}{6}\right|_0^1\\
  &=\frac{1}{6}
\end{align*}

This is reassuring since we would be surprised if the volume under a surface
were different if we sum in a different direction! In general, it is
possible to reverse the order of integration as long as the integrand is
continuous over the whole domain of integration **including the boundary**.

## Non-square domains

The domain of integration need not always be a square. Consider the
following case:

<img src='./figures/d3.png' width=400>

In this case, $x$ varies from 0 to 1 and $y$ varies from 0 to $1-x$.

Choosing the former description, we can integrate a function, say
$f(x,y)=x^2y$ over this domain:

\begin{align*}
  \int_0^1\int_0^{1-x}f(x,y)\,\mathrm{d}y\,\mathrm{d}x&=\int_0^1\int_0^{1-x} x^2y\,\mathrm{d}y\,\mathrm{d}x\\
  &=\int_0^1\left. \frac{x^2y^2}{2}\right|_0^{1-x}\mathrm{d} x\\
  &=\int_0^1\frac{x^2(1-x)^2}{2}-0\mathrm{d} x\\
  &=\int_0^1\frac{x^2-2x^3+x^4}{2}\mathrm{d} x\\
  &=\left.\frac{x^3}{6}-\frac{x^4}{4}+\frac{x^5}{10}\right|^1_0\\
  &=\frac{1}{6}-\frac{1}{4}+\frac{1}{10}\\
  &=\frac{10}{60}-\frac{15}{60}+\frac{6}{60}\\
  &=\frac{1}{60}
\end{align*}

What if we wish to change the order in which we conduct the integrations?
Then we have $y$ varying from 0 to 1 and $x$ going from 0 to $1-y$:

\begin{align*}
  \int_0^1\int_0^{1-y} x^2y \mathrm{d} x\mathrm{d} y
  &=\int_0^1\left.\frac{x^3y}{3}\right|_0^{1-y}\mathrm{d} x\\
  &=\int_0^1\frac{(1-y)^3y}{3}\mathrm{d} x\\
  &=\int_0^1\frac{y-3y^2+3y^3-y^4}{3}\mathrm{d} x\\
  &=\left.\frac{y^2}{6}-\frac{y^3}{3}+\frac{y^4}{4}-\frac{y^5}{15}\right|_0^1\\
  &=\frac{1}{6}-\frac{1}{3}+\frac{1}{4}-\frac{1}{15}\\
  &=\frac{10-20+15-4}{60}\\
  &=\frac{1}{60}
\end{align*}

### Example

Evaluate:

\begin{align*}
  \int\int_R \left(2x^2+y\right)\mathrm{d}y \mathrm{d}x
\end{align*}

where $R$ is the region between $x=0$ and $x=1$ and the curves $y=x^2$ and
$y=x$.

\begin{align*}
  \int_0^1\int_{x^2}^x \left(2x^2+y\right)\mathrm{d} y\mathrm{d}x
  &=\int_0^1\left.\left(2x^2y+\frac{y^2}{2}\right|_{x^2}^x\right)\mathrm{d} x\\
  &=\int_0^1 \left(2x^3+\frac{x^2}{2}-2x^4-\frac{x^4}{2}\right)\mathrm{d} x\\
  &=\left[\frac{x^4}{2}+\frac{x^3}{6}-\frac{2x^5}{5}-\frac{x^5}{10}\right]^1_0\\
  &=\frac{1}{2}+\frac{1}{6}-\frac{2}{5}-\frac{1}{10}\\
  &=\frac{15}{30}+\frac{5}{30}-\frac{12}{30}-\frac{3}{30}\\
  &=\frac{5}{30}\\
  &=\frac{1}{6}
\end{align*}

## Polar coordinates

We can use multiple integrals to evaluate areas by simply integrating the
function $f(x,y)=1$ over the domain. For example, we could calculate the
area of the circle of radius by integrating the unit function over the following
domain and multiplying by four:

<img src='./figures/d4.png' width=400>

This results in the following integral:

\begin{align*}
  \int_0^a\int_0^{(a^2-y^2)^{\frac{1}{2}}}\,\mathrm{d}x\,\mathrm{d}y
\end{align*}

We could evaluate this integral directly but we obtain a fairly nasty
integrand for the second integral. Furthermore, it is obvious that polar
coordinates are far more natural  when considering a domain which is a
sector of a circle.

In polar coordinates, the domain of the integral becomes $0<r<a$,
$0<\theta<\pi/2$. But how do we integrate in polar coordinates?

In one dimension, the (Riemann) integral is evaluated as the limit $\delta
x\rightarrow 0$ the approximation of the area under the function by a sum
of rectangular strips of width $\delta x$. Similarly, in two dimensions, the
volume under the surface defined by $f(x,y)$ is approximated by a sum of
columns area $\delta A$ and the integral is found as the limit $\delta A
\rightarrow 0$. In Cartesian coordinates, the dimensions of each column are
$\delta x \times \delta y$ so the limit is achieved as $\delta x \rightarrow
0 $ and $\delta y \rightarrow 0$.

<img src='./figures/d5.png' width=400>

The area of the integral element is therefore $\delta x \delta y$.

In polar coordinates, the elemental areas are between lines of constant
radius and angle:

<img src='./figures/d6.png' width=400>

In the radial direction, the extent of the elemental area is clearly $\delta
r$, but what about in the azimuthal direction? $\delta\theta$ is an angle,
not a distance and the distance increases the further the element is from
the origin. The extent in the azimuthal direction is therefore
$r\delta\theta$. The elemental area is not rectangular, so what is the
area? It turns out that in the limit as $\delta\theta\rightarrow 0$ and
$\delta r\rightarrow 0$, the area becomes more and more rectangular, so the
area is, in fact, $r\delta\theta\delta r$.

Hence when changing from $x,y$ Cartesian coordinates to polar coordinates, we
replace $\,\mathrm{d}y\,\mathrm{d}x$ by $r\,\mathrm{d}r\,\mathrm{d}\theta$. We will naturally also need to make
use of the coordinate relations:

\begin{gather*}
  x=r\cos\theta\\
  y=r\sin\theta
\end{gather*}

So now we can evaluate the area of a circle of radius a by evaluating the
integral:

\begin{align*}
  \int_0^{2\pi}\int_0^a r\,\mathrm{d}r\,\mathrm{d}\theta &=
  \int_0^{2\pi}\left.\frac{r^2}{2}\right|_0^a\,\mathrm{d}\theta \\
  &= \int_0^{2\pi}\frac{a^2}{2}\,\mathrm{d}\theta \\
  &=\left.\frac{a^2}{2}\theta\right|_0^2\pi\\
   &=\pi a^2
\end{align*}

or, for region in the figure above, the integral would be

\begin{equation*}
  \int_0^{\pi/2}\int_0^a r\,\mathrm{d}r\,\mathrm{d}\theta =\frac{\pi a^2}{4}.
\end{equation*}


### Example

We noted previously that there is no closed form solution for the indefinite
integral:

\begin{equation*}
  \int e^{-x^2}\,\mathrm{d}x
\end{equation*}

However we can use a change of coordinates to evaluate the definite
integral:

\begin{equation*}
  I=\int_0^{\infty}e^{-x^2}\,\mathrm{d}x
\end{equation*}

This is a somewhat tricky manoeuvre. First we write:

\begin{equation*}
  I^2=\left(\int_0^{\infty}e^{-x^2}\,\mathrm{d}x\right)^2
\end{equation*}

However since the variable of integration is arbitrary, we could also
write:

\begin{equation*}
  I^2=\int_0^{\infty}e^{-x^2}\,\mathrm{d}x\int_0^{\infty}e^{-y^2}\,\mathrm{d}y
\end{equation*}

Now the two integrands are completely independent so can pass through the
integral signs without difficulty, so we have:

\begin{align*}
  I^2&=\int_0^{\infty}\int_0^{\infty}e^{-x^2}e^{-y^2}\,\mathrm{d}x\,\mathrm{d}y  
  &=\int_0^{\infty}\int_0^{\infty}e^{-(x^2+y^2)}\,\mathrm{d}x\,\mathrm{d}y  
\end{align*}

Next we change to polar coordinates and use the Pythagorean identity ($x^2+y^2=r^2$):

\begin{equation*}
  I^2=\int_0^{\frac{\pi}{2}}\int_0^\infty e^{-r^2}r\,\mathrm{d}r\,\mathrm{d}\theta
\end{equation*}

Now we have a much easier integral to perform. We let $u=r^2$ and note that:

\begin{equation*}
  \frac{\mathrm{d}u}{\mathrm{d}r}=2r
\end{equation*}

So:

\begin{align*}
  I^2&= \int_0^{\frac{\pi}{2}}\int_0^\infty -\frac{e^{-u}}{2}\,\mathrm{d}u\,\mathrm{d}\theta\\
  &=\int_0^{\frac{\pi}{2}}\left.-\frac{e^{-u}}{2}\right|_0^{\infty}\,\mathrm{d}\theta\\
  &=\int_0^{\frac{\pi}{2}}\frac{1}{2}\,\mathrm{d}\theta\\
  &=\frac{\pi}{4}
\end{align*}

Hence:

\begin{equation*}
  I=\frac{\sqrt{\pi}}{2}
\end{equation*}

## Arbitrary changes of coordinates

In principle we can change to any set of coordinates we choose. In two
dimensions we denote the usual Cartesian coordinate system $x$, $y$ and
write $u$, $v$ for some other Coordinate system.

In this case, the change of coordinates for the integral becomes:

\begin{align*}
  \mathrm{d}x\mathrm{d}y&=\left|J\right| \mathrm{d}u\mathrm{d}v\\ 
\end{align*}

where the vertical bars around the symbol $J$ indicate absolute value. The quantity $J$ is known as the Jacobian or Jacobian determinant and is given by:

\begin{align*}
  J
  &\equiv \left|
    \begin{matrix}
      \frac{\partial x}{\partial u} & \frac{\partial x}{\partial v}\\
      \frac{\partial y}{\partial u} & \frac{\partial y}{\partial v}\\
    \end{matrix}
  \right|\\
  &=\frac{\partial x}{\partial u}\frac{\partial y}{\partial v}-\frac{\partial x}{\partial v}\frac{\partial y}{\partial u}
\end{align*}

A shorthand notation for the Jacobian,

\begin{align*}
  J&\equiv \left|\frac{\partial (x,y)}{\partial (u,v)}\right| 
\end{align*}

is sometimes used.

### Polar coordinates revisited

Let's reconsider the transformation from Cartesian into Polar
coordinates:

\begin{align*}
  x&=r\cos\theta\\
  y&=r\sin\theta
\end{align*}

The elements of the Jacobian matrix are:

\begin{align*}
  \frac{\partial x}{\partial r}&=\cos\theta & \frac{\partial y}{\partial r}&=\sin\theta\\
  \frac{\partial x}{\partial \theta}&=-r\sin\theta & \frac{\partial y}{\partial\theta}&=r\cos\theta \\
\end{align*}

Hence:

\begin{align*}
  J&=\frac{\partial x}{\partial r}\frac{\partial y}{\partial \theta}-\frac{\partial x}{\partial \theta}\frac{\partial y}{\partial r}\\
  &=r\cos^2\theta+r\sin^2\theta\\
  &=r
\end{align*}

So, as we have already discovered:

\begin{align}
  \mathrm{d}x\mathrm{d}y=r\mathrm{d}r\mathrm{d}\theta
\end{align}

## Cylindrical polar coordinates

There are two natural ways to extend polar coordinates into three
dimensions. The first is to add $z$ as a Cartesian coordinate normal to the
polar plane. In this case, the Jacobian is given by:

\begin{equation*}
  J=\left|
    \begin{matrix}
      \frac{\partial x}{\partial r} & \frac{\partial x}{\partial \theta} & \frac{\partial x}{\partial z} \\
      \frac{\partial y}{\partial r} & \frac{\partial y}{\partial \theta} & \frac{\partial y}{\partial z} \\
      \frac{\partial z}{\partial r} & \frac{\partial z}{\partial \theta} & \frac{\partial z}{\partial z} \\
    \end{matrix}
    \right|
\end{equation*}

The transformation from cylindrical polar to Cartesian coordinates is given by:

\begin{align*}
  x&=r\cos\theta\\
  y&=r\sin\theta\\
  z&=z
\end{align*}

The elements of the Jacobian matrix are:

\begin{align*}
  \frac{\partial x}{\partial r}&=\cos\theta & \frac{\partial y}{\partial r}&=\sin\theta & \frac{\partial z}{\partial r}&=0\\
  \frac{\partial x}{\partial \theta}&=-r\sin\theta &   \frac{\partial y}{\partial \theta}&=r\cos\theta &\frac{\partial z}{\partial \theta}&=0\\
  \frac{\partial x}{\partial z}&=0 &  \frac{\partial y}{\partial z}&=0  &  \frac{\partial z}{\partial z}&=1
\end{align*}

Hence:

\begin{align*}
  J&=\left|
    \begin{matrix}
      \cos\theta & -r\sin\theta & 0 \\
      \sin\theta & r\cos\theta & 0 \\
      0 & 0 & 1
    \end{matrix}
  \right|\\
  &=1\times(r\cos^2\theta+r\sin^2\theta)\\
  &=r
\end{align*}

Hence:

\begin{equation*}
  \mathrm{d}x\mathrm{d}y\mathrm{d}z=r\mathrm{d}r\mathrm{d}\theta\mathrm{d}z
\end{equation*}

## Spherical polar coordinates

The other natural manner in which polar coordinates can be extended to three
dimensions is to introduce another angle, essentially a latitude, so that
points with a given $r$ lie on the surface of a sphere.

In this case the coordinates are $r$, the radius, $\theta$ the azimuth and
$\phi$, the inclination. The transformation from Spherical polar to
Cartesian coordinates is given by:

\begin{align*}
  x&=r\cos\theta\sin\phi\\
  y&=r\sin\theta\sin\phi\\
  z&=r\cos\phi
\end{align*}

where $0\leq\theta\leq2\pi$ and $0\leq\phi\leq\pi$ with $\phi=0$ being the North Pole, $\phi=\pi/2$ being the equator, and $\phi=\pi$ being the South Pole. This time, the change of coordinates and its derivatives are:

\begin{align*}
  x&=r\cos\theta\sin\phi& y&=r\sin\theta\sin\phi & z&=r\cos\phi\\
  \frac{\partial x}{\partial r}&=\cos\theta\sin\phi & \frac{\partial y}{\partial r}&=\sin\theta\sin\phi & \frac{\partial z}{\partial r}&=\cos\phi\\
  \frac{\partial x}{\partial \theta}&=-r\sin\theta\sin\phi &   \frac{\partial y}{\partial \theta}&=r\cos\theta\sin\phi &\frac{\partial z}{\partial \theta}&=0\\
  \frac{\partial x}{\partial \theta}&=r\cos\theta\cos\phi &  \frac{\partial y}{\partial \phi}&=r\sin\theta\cos\phi  &  \frac{\partial z}{\partial \phi}&=-r\sin\phi
\end{align*}

So:

\begin{align*}
  J&=\left|
    \begin{matrix}
      \frac{\partial x}{\partial r} & \frac{\partial x}{\partial \theta} & \frac{\partial x}{\partial \phi} \\
      \frac{\partial y}{\partial r} & \frac{\partial y}{\partial \theta} & \frac{\partial y}{\partial \phi} \\
      \frac{\partial z}{\partial r} & \frac{\partial z}{\partial \theta} & \frac{\partial z}{\partial \phi} \\
    \end{matrix}
  \right|\\
  &=\left|
    \begin{matrix}
      \cos\theta\sin\phi & -r\sin\theta\sin\phi & r\cos\theta\cos\phi \\
      \sin\theta\sin\phi & r\cos\theta\sin\phi & r\sin\theta\cos\phi \\
      \cos\phi & 0 & -r\sin\phi \\
    \end{matrix}
  \right|\\
  &=\left|\cos\phi(-r^2\sin^2\theta\sin\phi\cos\phi-r^2\cos^2\theta\sin\phi\cos\phi)
  -r\sin\phi(r\cos^2\theta\sin^2\phi+r\sin^2\theta\sin^2\phi)\right|\\
  &=\left|-r^2(\sin\phi\cos^2\phi+\sin^3\phi)\right|\\
  &=r^2\sin\phi
\end{align*}

Hence:

\begin{equation*}
  \mathrm{d}x\mathrm{d}y\mathrm{d}z=r^2\sin\phi\mathrm{d}r\mathrm{d}\phi\mathrm{d}\theta
\end{equation*}

So the volume of the object can be calculated from 

\begin{equation*}
    V = \iiint_V \mathrm{d}x\mathrm{d}y\mathrm{d}z
\end{equation*}

using Cartesian coordinates, or 

\begin{equation*}
    V = \iiint_V r^2\sin\phi\mathrm{d}r\mathrm{d}\phi\mathrm{d}\theta
\end{equation*}

using Spherical polar coordinates.

### Example:

Consider a spherical planet and imagine that the density of this planet is equal to $\rho_i$ in the inner core and $\rho_0$ in the outer crust. The radius of the core is $b$ and the total radius of the planet is $a$.
 
 (a) The total volume of the planet can be expressed as
 
 \begin{equation*}
  V=\int_{\theta=0}^{2\pi}\int_{\phi=0}^{\phi=\pi}\int_{r=0}^{a} r^2\sin(\phi)\mathrm{d}r\mathrm{d}\phi\mathrm{d}\theta.
 \end{equation*}
 
 Evaluate this integral to find the total volume of the planet.
 
 (b) Now form and evaluate an integral for the total mass of the planet. Recall that mass is (density $\times$ volume).
 
 (c) What is the average density of the planet?

#### Solution:

(a) We have that the volume $V$ is given by

\begin{align*}
 V&=\int_0^{2\pi}\int_{\pi}\left[\frac{r^3}{3}\right]_{0}^{a}\sin\phi\mathrm{d}\phi\mathrm{d}\theta\\
  &=\frac{a^3}{3}\int_0^{2\pi}\int_0^{\pi}\sin\phi\mathrm{d}\phi\mathrm{d}\theta\\
  &=\frac{a^3}{3}\int_{0}^{2\pi}\left[-\cos\phi\right]_{0}^{\pi}\mathrm{d}\theta\\
  &=\frac{2a^3}{3}\int_{0}^{2\pi}\mathrm{d}\theta=\frac{4\pi a^3}{3}.
\end{align*}

(b) The total mass $M$ of the planet will be given by

\begin{equation*}
 M=\rho_i\int_{0}^{2\pi}\int_{0}^{\pi}\int_{0}^{b} r^2\sin(\phi)\mathrm{d}r\mathrm{d}\phi\mathrm{d}\theta+\rho_0\int_{0}^{2\pi}\int_{0}^{\pi}\int_{b}^{a} r^2\sin(\phi)\mathrm{d}r\mathrm{d}\phi\mathrm{d}\theta.
\end{equation*}

Using the result from part (a), we then get

\begin{align*}
 M=\frac{4\pi b^3}{3}\rho_i+\frac{4\pi (a^3-b^3)}{3}\rho_0=\frac{4\pi}{3}\left[\rho_i b^3+\rho_0(a^3-b^3)\right].
\end{align*}

(c) The average density $<\rho>$ will be given by $M/V$ which gives

\begin{equation*}
 <\rho>=\frac{\frac{4\pi}{3}\left[\rho_i b^3+\rho_0(a^3-b^3)\right]}{\frac{4\pi a^3}{3}}=\rho_i\left(\frac{b}{a}\right)^3+\rho_0\left[1-\left(\frac{b}{a}\right)^3\right].
\end{equation*}

## Other spherical polar notational conventions

Unfortunately, there is no 'standard' notion for how spherical polar coordinates are presented. If you look in the literature and through text books you'll likely run into a few different forms. Two of the other common notations are outlined below.

#### Common mathematician notation

\begin{align*}
 x&=r\cos\phi\sin\theta,\\
 y&=r\sin\phi\sin\theta,\\
 z&=r\cos\theta,
\end{align*}

where $0\leq\phi\leq2\pi$, $0\leq\theta\leq\pi$. Here we've simply swapped $\theta$ and $\phi$ notation, $\phi$ is now the azimuthal angle, and $\theta$ the inclination angle. Basically, when encountering a problem in spherical polars, be careful to check what $\theta$ and $\phi$ mean!

#### Atmospheric physicists notation

\begin{align*}
 x&=r\cos\phi\cos\theta,\\
 y&=r\sin\phi\cos\theta,\\
 z&=r\sin\theta,
\end{align*}

where $0\leq\phi\leq2\pi$, $-\pi/2\leq\theta\leq\pi/2$. Now, this one's a bit more different - $\theta=-\pi/2$ now represents the south pole, $\theta=0$ the equator and $\theta=\pi/2$ the north pole. Note you may also see this notation with the $\theta$ and $\phi$ swapped.
